```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Configuración
R = 1  # Radio del cilindro
lambda_max = 4 * np.pi  # Parámetro máximo
n_points = 1000

# Crear array del parámetro
lambda_vals = np.linspace(0, lambda_max, n_points)

# Caso 1: Línea vertical (ω = 0, v_z ≠ 0)
print("Caso 1: Línea vertical")
v_z1, omega1 = 0.5, 0
z1 = 0 + v_z1 * lambda_vals
phi1 = 0 + omega1 * lambda_vals
x1 = R * np.cos(phi1)
y1 = R * np.sin(phi1)

# Caso 2: Círculo horizontal (v_z = 0, ω ≠ 0)
print("Caso 2: Círculo horizontal")
v_z2, omega2 = 0, 1
z2 = 0 + v_z2 * lambda_vals
phi2 = 0 + omega2 * lambda_vals
x2 = R * np.cos(phi2)
y2 = R * np.sin(phi2)

# Caso 3: Hélice (v_z ≠ 0, ω ≠ 0)
print("Caso 3: Hélice")
v_z3, omega3 = 0.3, 2
z3 = 0 + v_z3 * lambda_vals
phi3 = 0 + omega3 * lambda_vals
x3 = R * np.cos(phi3)
y3 = R * np.sin(phi3)

# Visualización 3D
fig = plt.figure(figsize=(15, 5))

# Gráfico 3D
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot(x1, y1, z1, 'r-', linewidth=2, label='Línea vertical')
ax1.set_title('Caso 1: Línea Vertical\n(ω = 0, v_z ≠ 0)')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.legend()

ax2 = fig.add_subplot(132, projection='3d')
ax2.plot(x2, y2, z2, 'g-', linewidth=2, label='Círculo horizontal')
ax2.set_title('Caso 2: Círculo Horizontal\n(v_z = 0, ω ≠ 0)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.legend()

ax3 = fig.add_subplot(133, projection='3d')
ax3.plot(x3, y3, z3, 'b-', linewidth=2, label='Hélice')
ax3.set_title('Caso 3: Hélice\n(v_z ≠ 0, ω ≠ 0)')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('Z')
ax3.legend()

plt.tight_layout()
plt.show()

# Visualización del plano desarrollado
fig2, (ax4, ax5, ax6) = plt.subplots(1, 3, figsize=(15, 5))

# En el plano desarrollado: x = R*phi, y = z
ax4.plot(R * phi1, z1, 'r-', linewidth=2)
ax4.set_title('Plano Desarrollado: Línea Vertical')
ax4.set_xlabel('R·φ')
ax4.set_ylabel('z')
ax4.grid(True)
ax4.axis('equal')

ax5.plot(R * phi2, z2, 'g-', linewidth=2)
ax5.set_title('Plano Desarrollado: Círculo Horizontal')
ax5.set_xlabel('R·φ')
ax5.set_ylabel('z')
ax5.grid(True)
ax5.axis('equal')

ax6.plot(R * phi3, z3, 'b-', linewidth=2)
ax6.set_title('Plano Desarrollado: Hélice')
ax6.set_xlabel('R·φ')
ax6.set_ylabel('z')
ax6.grid(True)
ax6.axis('equal')

plt.tight_layout()
plt.show()

# Visualización comparativa de los tres casos
fig3 = plt.figure(figsize=(10, 8))
ax7 = fig3.add_subplot(111, projection='3d')

# Plot todos juntos
ax7.plot(x1, y1, z1, 'r-', linewidth=3, label='Línea vertical (ω=0)')
ax7.plot(x2, y2, z2, 'g-', linewidth=3, label='Círculo horizontal (v_z=0)')
ax7.plot(x3, y3, z3, 'b-', linewidth=3, label='Hélice (ambos ≠ 0)')

ax7.set_title('Todas las Geodésicas en el Cilindro')
ax7.set_xlabel('X')
ax7.set_ylabel('Y')
ax7.set_zlabel('Z')
ax7.legend()

# Añadir el cilindro transparente para referencia
phi_cyl = np.linspace(0, 2*np.pi, 50)
z_cyl = np.linspace(-2, 2, 50)
Phi_cyl, Z_cyl = np.meshgrid(phi_cyl, z_cyl)
X_cyl = R * np.cos(Phi_cyl)
Y_cyl = R * np.sin(Phi_cyl)

ax7.plot_surface(X_cyl, Y_cyl, Z_cyl, alpha=0.1, color='gray')

plt.show()

# Información numérica
print("\n" + "="*50)
print("INFORMACIÓN DE LAS GEODÉSICAS")
print("="*50)
print(f"Caso 1 - Línea vertical:")
print(f"  v_z = {v_z1:.2f}, ω = {omega1:.2f}")
print(f"  Distancia vertical recorrida: {z1[-1]:.2f}")

print(f"\nCaso 2 - Círculo horizontal:")
print(f"  v_z = {v_z2:.2f}, ω = {omega2:.2f}")
print(f"  Vueltas completas: {phi2[-1]/(2*np.pi):.2f}")

print(f"\nCaso 3 - Hélice:")
print(f"  v_z = {v_z3:.2f}, ω = {omega3:.2f}")
print(f"  Paso de la hélice: {2*np.pi*v_z3/omega3:.2f}")
print(f"  Vueltas completas: {phi3[-1]/(2*np.pi):.2f}")
print(f"  Distancia vertical: {z3[-1]:.2f}")
```
