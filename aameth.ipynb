```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft

# Definir la funcin peridica f()
def f(theta):
    return (theta**2 - np.pi**2)**2

# Definir el nmero de trminos mximos en la serie de Fourier
N_max = 5

# Valores de 
theta = np.linspace(-np.pi, np.pi, 1000)

# Calcular la serie de Fourier y aproximar la funcin para diferentes valores de N
plt.figure(figsize=(10, 6))
plt.plot(theta, f(theta), label='f()', color='blue')  # Graficar la funcin original

for N in range(1, N_max+1):
    # Calcular los coeficientes de la serie de Fourier
    c = fft(f(theta))
    
    # Conservar solo los primeros N trminos
    c[N+1:-N] = 0
    
    # Aproximar la funcin f() utilizando la serie de Fourier truncada
    f_approx = ifft(c).real
    
    # Graficar la serie de Fourier aproximada
    plt.plot(theta, f_approx, linestyle='--')

plt.xlabel('')
plt.ylabel('f()')
plt.title('Series de Fourier para f()')
plt.legend()
plt.show()
```
