## Números de fibonacci

```python
limit = int(input())
f1 , f2 = 1 , 1
while f1 <= limit: #
    print(f1)
    f1 , f2 = f2 , f1 + f2 #reemplazar el valor de f1 y f2
```

## Números Catalán

```python
#Si queremos guardar todos los valores en una lista
lista = []
limit0 = int(input())
c1 , c2 = 1 , 1
n = 0
while c1 <= limit0:
    #print(c1)
    lista.append(c1)
    c1 , c2 = ((4*n+2)/(n+2))*c1 , c1
    n += 1
print(lista)
```

```python
import numpy as np
```

```python
theta0 = 45*np.pi/180

r = [0,9.3]
rf = [2.1,0]
v = [10*np.cos(theta0),-10*np.sin(theta0)]
a = [0,-9.8]

t1 = (v[1]+np.sqrt(v[1]**2-4*0.5*a[1]*r[1]))/(-a[1]) #La respuesta es t1
t2 = (v[1]-np.sqrt(v[1]**2-4*0.5*a[1]*r[1]))/(-a[1])
print(t1,t2)

x = v[0]*t1
print(x)
```

```python
a0 = (rf[0]**2)*(-a[1])/200
newtan0 = (-rf[0]+np.sqrt(rf[0]**2-4*a0*(a0-r[1])))/(2*a0) #este es
newtan1 = (-rf[0]-np.sqrt(rf[0]**2-4*a0*(a0-r[1])))/(2*a0)
print(newtan0,newtan1)
factor = 180/np.pi
tanfin0, tanfin1 = np.arctan(newtan0), np.arctan(newtan1) 
print(tanfin0*factor,tanfin1*factor)
```
