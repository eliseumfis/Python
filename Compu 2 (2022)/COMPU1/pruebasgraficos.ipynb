```python
h = float(input("Altura de la torre: "))
g = 9.81
t0 = float(input("tiempo inicial"))
#tf = float(input("tiempo final"))
tf = np.sqrt(2*h/g)

x = np.arange(t0,tf,0.1)

y = h - .5 * g * x ** 2
print(y)
```
