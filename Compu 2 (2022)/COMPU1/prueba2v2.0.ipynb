```python
def media(arreglo,fila):
    m = arreglo[fila-1,]
    s = 0
    for i in range(len(m)):
        s += m[i]
    d = len(m)
    return(s/d)

def desviacion(arreglo,fila):					
	m = arreglo[fila-1,]
	s = 0
	for r in range(len(m)):             
		s += (m[r]-media(arreglo,fila))**2
	return(np.sqrt(s/(len(m)-1)))
```

```python
matriz = np.asarray([dia1,dia2,dia3,dia4,dia5,dia6,dia7,dia8,dia9,dia10,dia11,dia12]); print(matriz)
```

```python
print(media(matriz,1))
media0=[]
desviacion0=[]
for i in range(1,13):
    media0.append(media(matriz,i))
    desviacion0.append(desviacion(matriz,i))
print(media0,desviacion0)
```

```python
def media1(lista):
    return(sum(lista)/len(lista))

def desviacionestandar(lista):
    s = 0
    media = sum(lista)/len(lista)
    for i in lista:
        s += (i - media)**2
    d = len(lista)-1
    return(np.sqrt(s/d))

T = temp
m,d=[],[]
for e in range(0,12):
    m.append(media1(T[int(e*8):int(e*8+8)]))
    d.append(desviacionestandar(T[int(e*8):int(e*8+8)]))
print(m,d)
```
