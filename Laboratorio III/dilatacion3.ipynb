```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def alpha(DeltaL,Li,Ti,Tf):
    return DeltaL/(Li*(Tf-Ti))
def error(teo,exp):
    return 100*np.abs(teo-exp)/teo
```

# **Aluminio**

```python
L1al=50;T1al=20 
DeltaLal=0.091;T2al=98
DeltaTAl=T2al-T1al
ateoAl=23e-6
alphaAl=alpha(DeltaLal,L1al,T1al,T2al)
```

```python
print("El valor de la constante de dilatación lineal para el Aluminio es:",alpha(DeltaLal,L1al,T1al,T2al),", con un error del :",error(ateoAl,alpha(DeltaLal,L1al,T1al,T2al)),"con respecto al valor teórico.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función por partes
def f(x):
    if x < 4:
        return 0.5 * x**2 + 2
    elif 4 < x < 10:
        return 2 * x + 2
    else:
        return 7

# Vectorizar la función para poder aplicarla a arrays
f_vec = np.vectorize(f)

# Crear los intervalos para cada parte de la función
x1 = np.linspace(-5, 4, 500)  # Primera parte: x < 4
x2 = np.linspace(4, 10, 500)  # Segunda parte: 4 < x < 10
x3 = np.linspace(10, 15, 500) # Tercera parte: x >= 10

# Calcular los valores y para cada intervalo
y1 = f_vec(x1)
y2 = f_vec(x2)
y3 = f_vec(x3)

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Graficar cada parte de la función
plt.plot(x1, y1, 'b-', label=r'$0.5x^2 + 2$ (x < 4)')
plt.plot(x2, y2, 'r-', label=r'$2x + 2$ (4 < x < 10)')
plt.plot(x3, y3, 'g-', label='7 (x ≥ 10)')

# Marcar los puntos de discontinuidad (si los hay)
plt.plot(4, f_vec(4), 'bo', markersize=8)  # Punto en x=4
plt.plot(4, 2*4 + 2, 'ro', markersize=8)   # Punto en x=4 (límite derecho)
plt.plot(10, 2*10 + 2, 'ro', markersize=8) # Punto en x=10 (límite izquierdo)
plt.plot(10, 7, 'go', markersize=8)        # Punto en x=10

# Configuraciones adicionales del gráfico
plt.title('Gráfico de la función por partes', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.xlim(-5, 15)
plt.ylim(0, 25)

# Mostrar el gráfico
plt.show()
```
