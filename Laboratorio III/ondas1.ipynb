```python
import numpy as np

# Lista de velocidades medidas (v_i)
velocidades = [342.1, 342.0, 340.5]

# Lista de incertidumbres correspondientes (Δv_i)
incertidumbres = [1.1, 2.0, 3.0]

# Convertimos a arrays para vectorizar operaciones
v = np.array(velocidades)
dv = np.array(incertidumbres)

# Calculamos los pesos como el inverso del cuadrado de las incertidumbres
pesos = 1 / dv**2

# Promedio ponderado
v_prom = np.sum(v * pesos) / np.sum(pesos)

# Incertidumbre del promedio ponderado
dv_prom = np.sqrt(1 / np.sum(pesos))

# Resultado
print(f"Promedio ponderado de velocidades: {v_prom:.3f} ± {dv_prom:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes físicas arbitrarias
G = 1      # Constante gravitacional
M = 1      # Masa del cuerpo central
m = 1      # Masa de la partícula
L = 2      # Momento angular

# Rango de radios
r = np.linspace(0.1, 10, 500)

# Potencial gravitacional
V_grav = -G * M * m / r

# Término centrífugo
V_cent = L**2 / (2 * m * r**2)

# Potencial efectivo
V_eff = V_grav + V_cent

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(r, V_grav, label='Potencial gravitacional $V(r)$', linestyle='--')
plt.plot(r, V_cent, label='Término centrífugo', linestyle=':')
plt.plot(r, V_eff, label='Potencial efectivo $V_{efectivo}(r)$', linewidth=2)
plt.xlabel('Radio $r$')
plt.ylabel('Potencial')
plt.title('Potencial efectivo $V_{efectivo}(r)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
