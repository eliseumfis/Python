```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Nuevos datos: rayo pasando del acrílico al aire
ang_incc = np.array([5,10,15,20,25,30,35,40])
ang_refracc = np.array([8,12,25,32,41,49,60,73])

# Paso 1: Calcular senos de los ángulos
sen_theta_i = np.sin(np.radians(ang_incc))     # Incidencia desde el acrílico
sen_theta_t = np.sin(np.radians(ang_refracc))  # Refracción en el aire

# Paso 2: Calcular índice de refracción relativo n_acrilico / n_aire = senθ_t / senθ_i => n_acrilico = senθ_t / senθ_i
n_relativo_array = sen_theta_t / sen_theta_i
n_prom = np.mean(n_relativo_array)

# Paso 3: Ajuste lineal de sin(θi) vs sin(θt)
slope, intercept, r_value, p_value, std_err = linregress(sen_theta_i, sen_theta_t)
n_estimada_ajuste = slope  # Aquí ya es n2/n1 directamente

# Paso 4: Gráfica
plt.figure(figsize=(10,7))
plt.plot(sen_theta_i, sen_theta_t, 'o', label='Datos medidos',color="black")
plt.plot(sen_theta_i, slope * sen_theta_i + intercept, 'r--', label=f'Ajuste lineal: $n_2$ ≈ {slope:.3f}')
plt.xlabel(r'$\sin(\theta_i)$', fontsize=17)
plt.ylabel(r'$\sin(\theta_t)$', fontsize=17)
plt.title('Ley de Snell (Acrílico → Aire)',fontsize=22)
plt.grid(True)
plt.legend(fontsize=15)
plt.tight_layout()
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
plt.show()

# Paso 5: Imprimir resultados
print("Índices de refracción relativos (n_acrilico):")
print(np.round(n_relativo_array, 3))

print(f"\nÍndice de refracción promedio: {n_prom:.3f}")
print(f"Índice de refracción estimado por regresión lineal: {slope:.3f}")
print(f"Intercepto de la regresión: {intercept:.5f}")
print(f"Coeficiente de correlación R²: {r_value**2:.4f}")
```
