```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
k=6.235256926e19 # [kg/m]
Msol=1.98e30 # [kg]
Mo=3.7e6 * Msol
kk=k*3.08e16
```

```python
def M(r):
    return (kk*r - Mo)/Msol
```

```python
rfit=np.linspace(0.01,1000,5000)
```

```python
mfits=[]
for c in rfit:
    mfits.append(M(c))
    
```

```python
logimasas=np.log10(mfits)
```

```python
logires=np.log10(rfit)
```

```python
plt.plot(logires,logimasas)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
k = 6.23e19  # kg/m
M0_solar = 3.7e6  # en masas solares
M_sun_kg = 1.989e30  # masa solar en kg
pc_to_m = 3.086e16  # parsec a metros

# Convertir M0 a kilogramos
M0_kg = M0_solar * M_sun_kg

# Rango de r en parsecs (logartmicamente espaciado)
r_pc = np.logspace(-2, 3, 500)  # de 0.01 pc a 1 kpc en parsecs

# Convertir r a metros
r_m = r_pc * pc_to_m

# Calcular M_r en kg y luego convertir a masas solares
Mr_kg = k * r_m + M0_kg
Mr_solar = Mr_kg / M_sun_kg

# Graficar log10(M_r) vs log10(r)
plt.figure(figsize=(8, 6))
plt.plot(np.log10(r_pc), np.log10(Mr_solar), label=r"$\log_{10}(M_r)$ vs $\log_{10}(r)$", color='blue')
plt.xlabel(r"$\log_{10}(r) \, [\mathrm{pc}]$", fontsize=14)
plt.ylabel(r"$\log_{10}(M_r) \, [M_\odot]$", fontsize=14)
plt.title(r"Relacin entre $\log_{10}(M_r)$ y $\log_{10}(r)$", fontsize=16)
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
G = 6.674e-11  # Constante gravitacional en m^3 kg^-1 s^-2
k = 6.23e19  # kg/m
M0_solar = 3.7e6  # en masas solares
M_sun_kg = 1.989e30  # masa solar en kg
pc_to_m = 3.086e16  # parsec a metros

# Convertir M0 a kilogramos
M0_kg = M0_solar * M_sun_kg

# Rango de r en parsecs (logartmicamente espaciado)
r_pc = np.logspace(-2, 3, 500)  # de 0.01 pc a 1 kpc en parsecs

# Convertir r a metros
r_m = r_pc * pc_to_m

# Calcular v(r) en m/s usando la frmula dada
v_m_per_s = np.sqrt(G * (k + M0_kg / r_m))

# Convertir v(r) a km/s
v_km_per_s = v_m_per_s / 1000

# Graficar v(r) vs log10(r)
plt.figure(figsize=(8, 6))
plt.plot(np.log10(r_pc), v_km_per_s, label=r"$v(r)$ vs $\log_{10}(r)$", color='green')
plt.xlabel(r"$\log_{10}(r) \, [\mathrm{pc}]$", fontsize=14)
plt.ylabel(r"$v(r) \, [\mathrm{km/s}]$", fontsize=14)
plt.title(r"Relacin entre $v(r)$ y $\log_{10}(r)$", fontsize=16)
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```
