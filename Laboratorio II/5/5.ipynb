```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del circuito
R = 1485  # Resistencia en ohmios
C = 0.215e-6  # Capacitancia en faradios
tau = R * C  # Constante de tiempo del circuito

# Tiempo de simulacin
t = np.linspace(0, 0.01, 1000)  # 10 ms con 1000 puntos

# Seal de carga y descarga
V_in = 5  # Voltaje de entrada en voltios
period = 0.004  # Periodo de la seal en segundos
duty_cycle = 0.5  # Ciclo til (50%)

# Generar una onda cuadrada como entrada
signal = ((t % period) < (duty_cycle * period)).astype(float) * V_in

# Respuesta del condensador
V_out = np.zeros_like(t)
for i in range(1, len(t)):
    if signal[i] > 0:  # Carga
        V_out[i] = V_out[i-1] + (signal[i] - V_out[i-1]) * (t[i] - t[i-1]) / tau
    else:  # Descarga
        V_out[i] = V_out[i-1] * np.exp(-(t[i] - t[i-1]) / tau)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(t, V_out, label="Voltaje en el condensador", color="red")
plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")
plt.title("Carga y descarga de un condensador")
plt.grid()
plt.legend()
plt.show()
```
