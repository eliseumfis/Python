```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
R = 1485  # Resistencia en ohmios
C = 0.215e-6  # Capacitancia en faradios
```

```python
# Datos
# Datos
V_0 = np.array([39.810, 39.910, 39.933, 39.913, 39.852, 39.957, 39.942, 39.918, 39.864,
                39.910, 39.930, 39.886, 39.928, 39.858, 39.942, 39.904, 39.928, 39.949,
                39.925, 39.943, 39.907, 39.873, 39.870, 39.945, 39.917, 39.946, 39.981])  # Voltios
V_C = np.array([39.012, 38.137, 37.002, 34.242, 32.570, 31.028, 28.168, 25.477, 21.025,
                16.422, 12.503, 10.612, 9.202, 8.065, 6.966, 6.109, 5.448, 4.805, 4.289,
                3.656, 3.291, 3.115, 2.993, 2.825, 2.706, 2.566, 2.380])  # Voltios
delta_t = np.array([359.848, 340.909, 321.970, 303.030, 344.697, 306.818, 251.894, 255.682,
                    187.500, 172.348, 145.833, 115.530, 104.167, 98.485, 79.545, 68.182,
                    68.182, 53.977, 53.977, 45.455, 41.667, 39.773, 37.879, 35.227,
                    29.924, 32.576, 29.167]) #* 1e-6  # Convertir a segundos
f = np.array([50, 150, 200, 300, 350, 400, 500, 600, 800, 1100, 1500, 1800, 2100, 2400,
               2800, 3200, 3600, 4100, 4600, 5400, 6000, 6300, 6600, 7000, 7300, 7700, 8300])  # Hz
```

```python
print(np.size(V_0))
```

```python
# Cálculos
omega = 2 * np.pi * f  # Frecuencia angular
phi = 2 * np.pi *f * delta_t  # Desfase en radianes
```

```python
tfitt=np.linspace(0,1000,27)#* 1e-6
ffitt=np.linspace(50,8300,27)
vcfitt=np.linspace(39.012,4.289,27)
v0fitt=np.linspace(39.810,39.981,27)

omegafitt=2*np.pi*ffitt
phifitt=2*np.pi*ffitt*tfitt
```

```python
tfit=np.linspace(0,5000,250)#* 1e-6
ffit=np.linspace(50,8300,250)
vcfit=np.linspace(39.012,4.289,250)
v0fit=np.linspace(39.810,39.981,250)

omegafit=2*np.pi*ffit
phifit=2*np.pi*ffit*tfit
```

```python
tfit2=np.linspace(0,5000,250)#* 1e-6
ffit2=np.linspace(50,8300,250)
vcfit2=np.linspace(39.012,4.289,250)
v0fit2=np.linspace(39.810,39.981,250)

omegafit2=2*np.pi*ffit2
phifit2=2*np.pi*ffit2*tfit2
```

```python
def voltcondensador(w,t,o):
    return (np.mean(V_0)/np.sqrt(1485**2 + 1/(w*C)**2))*((1/(w*C))*np.sin(w*t+o))
def voltfuente(w,t,o):
    return (np.mean(V_0)/np.sqrt(1485**2 + 1/(w*C)**2))*R*np.cos(w*t+o)+(np.mean(V_0)/np.sqrt(1485**2 + 1/(w*C)**2))*((1/(w*C))*np.sin(w*t+o))
```

```python
# Graficar
plt.figure(figsize=(10, 6))
plt.plot(tfit,abs(voltcondensador(omegafit,tfit,phifit)),color="black",label=r"$V_C(t)=I_0\sin(\omega t +\phi)/\omega C$")
plt.plot(tfit2,np.abs(voltfuente(omegafit2,tfit2,phifit2)),color="lightgray",label=r"$V(t)=I_0\cos(\omega t+\phi) + V_C$")
plt.title(r"Voltaje de fuente alterna y condensador",fontsize=17)
plt.ylabel("V [V]",fontsize=14.5);plt.xlabel(r"t [$\mu$s]",fontsize=14.5)
plt.legend(fontsize=14,loc=7)
plt.tick_params(axis='both', which='major', labelsize=12.5)  # Cambia el valor de labelsize según lo desees
plt.tight_layout()
```

```python
from scipy.stats import pearsonr
```

```python
V_c_div_V_0 = V_C / V_0
V_c_div_V_0_teorico=voltcondensador(omegafitt,tfitt,phifitt)/v0fitt
```

```python
print(V_c_div_V_0)
print(V_c_div_V_0_teorico)
```

```python
# Coeficientes de correlación de Pearson
corr_vc, _ = pearsonr(V_c_div_V_0, V_c_div_V_0_teorico)
#corr_phi, _ = pearsonr(phi, phi_teorico)

print(f"Coeficiente de correlación de Pearson para V_C(ω)/V_0: {corr_vc:.4f}")
#print(f"Coeficiente de correlación de Pearson para φ(ω): {corr_phi:.4f}")
```
