```python
import numpy as np

# Definir los valores de resistencias y sus errores
resistencias = [613, 747, 904, 1475]  # Ejemplo de resistencias en Ohms
errores_resistencias = [1, 1, 1, 1]  # Ejemplo de errores en Ohms

# Definir los voltajes de las fuentes y sus errores
voltajes = [7.5, 12]  # Ejemplo de voltajes en Voltios
errores_voltajes = [0.01, 0.01]  # Ejemplo de errores en Voltios

# Número de simulaciones
n_simulaciones = 100

# Función para calcular las corrientes según las leyes de Kirchhoff
def calcular_corrientes(resistencias, voltajes):
    # Aquí puedes implementar las ecuaciones de Kirchhoff para resolver el sistema
    # Por ejemplo, usando un sistema de ecuaciones lineales con numpy
    # Suponiendo que tienes las ecuaciones ya preparadas
    # Solo para ejemplo, devolvemos corrientes aleatorias
    return np.random.rand(len(resistencias))  # Devuelve un array de corrientes

# Arrays para almacenar los resultados
corrientes_simuladas = []

# Realizar 100 simulaciones
for _ in range(n_simulaciones):
    # Generar valores aleatorios dentro del rango de error permitido para resistencias y voltajes
    resistencias_random = [np.random.uniform(r - e, r + e) for r, e in zip(resistencias, errores_resistencias)]
    voltajes_random = [np.random.uniform(v - e, v + e) for v, e in zip(voltajes, errores_voltajes)]
    
    # Calcular las corrientes con estos valores
    corrientes = calcular_corrientes(resistencias_random, voltajes_random)
    
    # Guardar los resultados
    corrientes_simuladas.append(corrientes)

# Convertir a numpy array para facilidad de análisis
corrientes_simuladas = np.array(corrientes_simuladas)

# Calcular la media y la desviación estándar de las corrientes
corrientes_promedio = np.mean(corrientes_simuladas, axis=0)
desviacion_estandar = np.std(corrientes_simuladas, axis=0)

# Mostrar los resultados
for i, (promedio, desviacion) in enumerate(zip(corrientes_promedio, desviacion_estandar)):
    print(f"Corriente {i+1}: {promedio:.2f} A ± {desviacion:.2f} A")
```

```python
import numpy as np

# Valores de las resistencias
R1 = 100  # Ohmios
R2 = 200  # Ohmios
R3 = 300  # Ohmios
R4 = 400  # Ohmios

# Valores de los voltajes
V1 = 10  # Voltios
V2 = 5   # Voltios

# Sistema de ecuaciones a partir de las leyes de Kirchhoff
# I1 - I5 - I2 = 0  ->  ecuación 1
# I3 - I5 - I4 = 0  ->  ecuación 2
# V1 - I1*R1 - I2*R2 = 0  ->  ecuación 3
# V2 - I3*R3 - I4*R4 = 0  ->  ecuación 4
# V2 - V1 - I3*R3 + I2*R2 = 0  ->  ecuación 5

# Coeficientes de las corrientes I1, I2, I3, I4, I5
A = np.array([[1, -1, 0, 0, -1],  # Ecuación 1 (Ley de corrientes)
              [0, 0, 1, -1, -1],  # Ecuación 2 (Ley de corrientes)
              [R1, R2, 0, 0, 0],  # Ecuación 3 (Ley de mallas)
              [0, 0, R3, R4, 0],  # Ecuación 4 (Ley de mallas)
              [0, R2, -R3, 0, 0]])  # Ecuación 5 (Ley de mallas)

# Vector de resultados (los voltajes suministrados)
b = np.array([0, 0, V1, V2, V2 - V1])

# Resolver el sistema de ecuaciones
corrientes = np.linalg.solve(A, b)

# Mostrar los resultados
for i, corriente in enumerate(corrientes, 1):
    print(f"Corriente I{i}: {corriente:.2f} A")
```

```python
import numpy as np

# Valores iniciales y sus errores
R1_mean, R1_err = 613, 1
R2_mean, R2_err = 747, 1
R3_mean, R3_err = 904, 1
R4_mean, R4_err = 1475, 1
V1_mean, V1_err = 7.47, 0.01
V2_mean, V2_err = 11.91, 0.01

# Generar 100 valores aleatorios dentro de los rangos permitidos
R1_values = np.random.uniform(R1_mean - R1_err, R1_mean + R1_err, 100)
R2_values = np.random.uniform(R2_mean - R2_err, R2_mean + R2_err, 100)
R3_values = np.random.uniform(R3_mean - R3_err, R3_mean + R3_err, 100)
R4_values = np.random.uniform(R4_mean - R4_err, R4_mean + R4_err, 100)
V1_values = np.random.uniform(V1_mean - V1_err, V1_mean + V1_err, 100)
V2_values = np.random.uniform(V2_mean - V2_err, V2_mean + V2_err, 100)

# Función para calcular las corrientes aplicando las leyes de Kirchhoff
def calcular_corrientes(R1, R2, R3, R4, V1, V2):
    # Ejemplo de cálculos de corrientes según las ecuaciones (modificadas según el circuito)
    # Se pueden resolver usando álgebra lineal o métodos numéricos
    # Aquí debes agregar las ecuaciones del sistema basadas en las leyes de Kirchhoff
    I1 = (V1 - V2) / (R1 + R2)  # Ejemplo
    I2 = V1 / R3  # Ejemplo
    I3 = V2 / R4  # Ejemplo
    return I1, I2, I3

# Almacenar resultados
I1_results, I2_results, I3_results = [], [], []

for R1, R2, R3, R4, V1, V2 in zip(R1_values, R2_values, R3_values, R4_values, V1_values, V2_values):
    I1, I2, I3 = calcular_corrientes(R1, R2, R3, R4, V1, V2)
    I1_results.append(I1)
    I2_results.append(I2)
    I3_results.append(I3)

# Convertir a arrays para calcular media y desviación estándar
I1_results = np.array(I1_results)
I2_results = np.array(I2_results)
I3_results = np.array(I3_results)

# Calcular promedio y desviación estándar
I1_mean, I1_std = np.mean(I1_results), np.std(I1_results)
I2_mean, I2_std = np.mean(I2_results), np.std(I2_results)
I3_mean, I3_std = np.mean(I3_results), np.std(I3_results)

print(f"I1 promedio: {I1_mean:.2f} mA, desviación estándar: {I1_std:.2f} mA")
print(f"I2 promedio: {I2_mean:.2f} mA, desviación estándar: {I2_std:.2f} mA")
print(f"I3 promedio: {I3_mean:.2f} mA, desviación estándar: {I3_std:.2f} mA")
```
