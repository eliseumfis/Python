```python
import numpy as np

# Desviación estándar
sigma = desviacion_estandar_A

# Promedio de A12V_A
prom = np.mean(A12V_A)
print(f"El promedio es: {prom}")

# Definir la función chi-cuadrado
def chi_cuadrado(A12V_A, prom, sigma):
    chi2 = 0
    for i in range(len(A12V_A)):
        Oi = A12V_A[i]
        Vi = prom
        chi2 += ((Oi - Vi) ** 2) / (sigma ** 2)
    return chi2

# Calcular el valor de chi-cuadrado
chi2_value = chi_cuadrado(A12V_A, prom, sigma)
print(f"El valor de chi-cuadrado es: {chi2_value:.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
R12V = [
    0, 5, 10, 20, 50, 70, 90, 100, 150, 170,
    190, 200, 250, 300, 350, 400, 450, 500, 600, 700,
    800, 900, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500
]

A12V_A = [
    0.01596, 0.01586, 0.01576, 0.01554, 0.01495, 0.01459, 0.01423, 0.01407, 0.01328, 0.01299,
    0.01271, 0.01259, 0.01196, 0.01138, 0.01086, 0.01039, 0.00995, 0.00956, 0.00885, 0.00824,
    0.0077, 0.00724, 0.00682, 0.0053, 0.00435, 0.00367, 0.00318, 0.0028, 0.00251, 0.00227
]

# Desviación estándar
sigma = 0.00451

# Promedio de A12V_A
prom = np.mean(A12V_A)

# Función chi-cuadrado para cada resistencia
chi2_values = []

for i in range(len(R12V)):
    Oi = A12V_A[i]
    Vi = prom
    chi2 = ((Oi - Vi) ** 2) / (sigma ** 2)
    chi2_values.append(chi2)

# Generar la gráfica de Resistencia vs Chi-cuadrado
plt.figure(figsize=(10, 6))
plt.plot(R12V, chi2_values, marker='o', linestyle='-', color='b')
plt.xlabel('Resistencia (Ω)')
plt.ylabel(r'$\chi^2$')
plt.title('Gráfica de Resistencia vs Chi-cuadrado')
plt.grid(True)
plt.show()
```
