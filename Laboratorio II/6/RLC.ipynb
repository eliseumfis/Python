```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
MT=12.75
OSC=22.80
```

```python
# R=2090 [OHM]         L=35 mH            C=0.335 mF

Amplitudes=np.array([0.04,0.20,3.8,4.6,5.6,6,6.1,5.8,5.4,4.8,4.1,3.7,03.3,2.5,2.3,1.5,1,0.04,0.01,0.02]) # [mA]
Frecuencias=np.array([40,184,388,508,767,1068,1580,2293,3001,3945,5082,5698,6360,7750,8040,9750,11170,13250,17160,15400])# [Hz]
angulares=[]

    
```

```python
for k in range(len(Amplitudes)):
    angulares.append(2*np.pi*Frecuencias[k])
   
```

```python
print(angulares)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(angulares,Amplitudes,color="black",s=50,label=r"Datos experimentales")
plt.xlabel(r"$\omega$ [rad/s]",fontsize=15)
plt.ylabel(r"I [mA]",fontsize=15)
plt.scatter(9227.43,6.1,marker="*",color="crimson",s=150,label=r"$\omega_c=9227.43$ [rad/s]")

plt.axvline(x=9227.43, color="lightcoral", linestyle="--")
plt.title(r"Intensidad de corriente en funcin de las frecuencias",fontsize=18)

plt.grid()
plt.tick_params(axis='both', which='major', labelsize=13)  # Cambia el valor de labelsize segn lo desees
plt.legend(fontsize=15)
```
