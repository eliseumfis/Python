```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros del sistema
  # relacin de masas
M1 = 5  # masa de la estrella primaria
M2 = 0.2  # masa de la estrella secundaria
G = 1  # constante gravitacional simplificada
R = 1  # distancia entre las dos estrellas

# Crear la malla de puntos (espacio de coordenadas)
x = np.linspace(-3, 3, 500)
y = np.linspace(-3, 3, 500)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Funcin para calcular el potencial
def potencial(x, y, M, rx, ry):
    r = np.sqrt((x - rx)**2 + (y - ry)**2)
    return -M / r

# Puntos de las dos estrellas
x1, y1 = -R * (M2 / (M1 + M2)), 0  # Estrella primaria
x2, y2 = R * (M1 / (M1 + M2)), 0   # Estrella secundaria

# Calcular el potencial en cada punto de la malla
for i in range(len(x)):
    for j in range(len(y)):
        Z[i, j] = (potencial(X[i, j], Y[i, j], M1, x1, y1) + 
                   potencial(X[i, j], Y[i, j], M2, x2, y2))

# Graficar las curvas equipotenciales
plt.figure(figsize=(8, 8))
contour = plt.contour(X, Y, Z, levels=np.linspace(-5, 0, 6), cmap='RdBu')
plt.axvline(x=0, color='black', lw=1)
plt.axhline(y=0, color='black', lw=1)
plt.gca().set_aspect('equal', adjustable='box')

# Etiquetas de los puntos Lagrange
plt.text(-0.5, 0.5, '$L_1$', fontsize=12, color='purple')
plt.text(0.5, 0.5, '$L_2$', fontsize=12, color='green')
plt.text(1.0, 1.5, '$L_3$', fontsize=12, color='blue')

# Ttulos y etiquetas
plt.title('Lbulo de Roche')
plt.xlabel('x')
plt.ylabel('y')

# Mostrar leyenda de niveles
cbar = plt.colorbar(contour)
cbar.set_label('Potencial')

# Mostrar el grfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Funcin para calcular el potencial de cada estrella
def potencial(x, y, M, rx, ry):
    r = np.sqrt((x - rx)**2 + (y - ry)**2)
    return -M / r

# Funcin para calcular el campo gravitacional total en un punto (x, y)
def calcular_potencial(x, y, M1, M2, R):
    # Posicin de las dos estrellas en el eje X
    x1, y1 = -R * (M2 / (M1 + M2)), 0  # Estrella primaria
    x2, y2 = R * (M1 / (M1 + M2)), 0   # Estrella secundaria

    # Calcular el potencial total en el punto (x, y)
    V = (potencial(x, y, M1, x1, y1) + 
         potencial(x, y, M2, x2, y2))
    return V

# Funcin para generar el grfico del lbulo de Roche
def lobulo_roche(M1, M2, R, limite=5, puntos=500):
    # Crear la malla de puntos (espacio de coordenadas)
    x = np.linspace(-limite, limite, puntos)
    y = np.linspace(-limite, limite, puntos)
    X, Y = np.meshgrid(x, y)
    
    # Calcular el potencial total en cada punto de la malla
    Z = np.zeros_like(X)
    for i in range(len(x)):
        for j in range(len(y)):
            Z[i, j] = calcular_potencial(X[i, j], Y[i, j], M1, M2, R)

    # Graficar las curvas equipotenciales
    plt.figure(figsize=(8, 8))
    contour = plt.contour(X, Y, Z, levels=np.linspace(-5, 0, 6), cmap='RdBu')
    plt.axvline(x=0, color='black', lw=1)
    plt.axhline(y=0, color='black', lw=1)
    plt.gca().set_aspect('equal', adjustable='box')

    # Etiquetas de los puntos Lagrange (aproximados)
    plt.text(-0.5, 0.5, '$L_1$', fontsize=12, color='purple')
    plt.text(0.5, 0.5, '$L_2$', fontsize=12, color='green')
    plt.text(1.0, 1.5, '$L_3$', fontsize=12, color='blue')

    # Ttulos y etiquetas
    plt.title(f'Lbulo de Roche para M1 = {M1}, M2 = {M2}, R = {R}')
    plt.xlabel('x')
    plt.ylabel('y')

    # Mostrar leyenda de niveles
    cbar = plt.colorbar(contour)
    cbar.set_label('Potencial')

    # Mostrar el grfico
    plt.show()

# Parmetros generales (puedes cambiar estos valores)
M1 = 1   # Masa de la estrella primaria
M2 = 2 # Masa de la estrella secundaria
R = 1    # Distancia entre las dos estrellas

# Llamar a la funcin para graficar el lbulo de Roche
lobulo_roche(M1, M2, R)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Crear figura
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# -------------------------
# Panel 1: Roche Lobe Overflow (LMXB)
# -------------------------
ax = axes[0]
ax.set_title("Roche Lobe Overflow (LMXB)", fontsize=12, weight="bold")

# Dibujar estrella compaera (crculo grande)
companion = patches.Circle((-0.5, 0), 0.4, color="blue", alpha=0.8, label="Estrella compaera (baja masa)")
ax.add_patch(companion)

# Dibujar objeto compacto (crculo pequeo)
compact = patches.Circle((0.8, 0), 0.15, color="black", label="Agujero negro / NS")
ax.add_patch(compact)

# Dibujar flujo de gas (flechas)
for x in [-0.1, 0.0, 0.1]:
    ax.arrow(x, 0, 0.6, 0, head_width=0.08, head_length=0.1, fc="green", ec="green", alpha=0.7)

# Disco de acrecin (elipse alrededor del objeto compacto)
disk = patches.Ellipse((0.8, 0), 0.6, 0.3, fill=False, edgecolor="green", lw=2)
ax.add_patch(disk)

# -------------------------
# Panel 2: Wind Accretion (HMXB)
# -------------------------
ax = axes[1]
ax.set_title("Wind Accretion (HMXB)", fontsize=12, weight="bold")

# Dibujar estrella compaera masiva
massive = patches.Circle((-0.6, 0), 0.5, color="cyan", alpha=0.8, label="Estrella masiva (OB)")
ax.add_patch(massive)

# Dibujar objeto compacto
compact2 = patches.Circle((0.9, 0), 0.15, color="black", label="Agujero negro / NS")
ax.add_patch(compact2)

# Dibujar viento estelar (flechas radiales desde la estrella masiva hacia la derecha)
for y in [-0.3, -0.15, 0, 0.15, 0.3]:
    ax.arrow(-0.2, y, 0.9, 0, head_width=0.08, head_length=0.1, fc="green", ec="green", alpha=0.7)

# Disco pequeo alrededor del objeto compacto
disk2 = patches.Ellipse((0.9, 0), 0.4, 0.2, fill=False, edgecolor="green", lw=2)
ax.add_patch(disk2)

# -------------------------
# Estilo general
for ax in axes:
    ax.set_xlim(-1.2, 1.5)
    ax.set_ylim(-0.8, 0.8)
    ax.set_aspect("equal")
    ax.axis("off")

plt.tight_layout()
plt.show()
```
