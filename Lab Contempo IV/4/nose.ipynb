```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import matplotlib.patches as patches

# Configuracin de estilo para publicacin cientfica
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
plt.rcParams['legend.fontsize'] = 15
plt.rcParams['figure.dpi'] = 300

# Datos
thetas = np.array([14.0416667, 15.29166667, 17.375, 19.333333, 20.75, 21.4333333]) * np.pi / 180
sin_thetas = np.sin(thetas)
d = 1 / 600000  # 600 lneas/mm  600000 lneas/m
lambdas_exp = np.array([404.3790842, 439.35545966, 497.7073299, 551.7723631, 590.4850633, 609.0306481])
lambdas_teo = np.array([405, 436, 492, 546, 579, 623])
k = lambdas_teo * 10**(-9) / d
thetas_teo = np.arcsin(k)
sin_thetas_teo = np.sin(thetas_teo)

# Calcular errores porcentuales
errors = np.abs((lambdas_exp - lambdas_teo) / lambdas_teo) * 100

# Conversin nm  RGB
def wavelength_to_rgb(wavelength):
    gamma = 0.8
    intensity_max = 255
    factor = 0.0
    R = G = B = 0

    if 380 <= wavelength < 440:
        R = -(wavelength - 440) / (440 - 380)
        G = 0.0
        B = 1.0
    elif 440 <= wavelength < 490:
        R = 0.0
        G = (wavelength - 440) / (490 - 440)
        B = 1.0
    elif 490 <= wavelength < 510:
        R = 0.0
        G = 1.0
        B = -(wavelength - 510) / (510 - 490)
    elif 510 <= wavelength < 580:
        R = (wavelength - 510) / (580 - 510)
        G = 1.0
        B = 0.0
    elif 580 <= wavelength < 645:
        R = 1.0
        G = -(wavelength - 645) / (645 - 580)
        B = 0.0
    elif 645 <= wavelength <= 780:
        R = 1.0
        G = 0.0
        B = 0.0
    
    if 380 <= wavelength < 420:
        factor = 0.3 + 0.7*(wavelength - 380) / (420 - 380)
    elif 420 <= wavelength < 701:
        factor = 1.0
    elif 701 <= wavelength <= 780:
        factor = 0.3 + 0.7*(780 - wavelength) / (780 - 700)
    
    R = round(intensity_max * (R * factor)**gamma)
    G = round(intensity_max * (G * factor)**gamma)
    B = round(intensity_max * (B * factor)**gamma)

    return (R/255, G/255, B/255)

# Crear gradiente del espectro visible
wavelengths = np.linspace(380, 780, 1000)
colors = [wavelength_to_rgb(w) for w in wavelengths]

# Crear figura con dos subplots
fig = plt.figure(figsize=(12, 11))
gs = fig.add_gridspec(2, 1, height_ratios=[3, 1], hspace=0.3)

# Primer subplot: datos principales
ax1 = fig.add_subplot(gs[0])

# Imagen de fondo (espectro continuo)
gradient = np.array([colors])  # banda horizontal
ax1.imshow(gradient, extent=[380, 780, ax1.get_ylim()[0], ax1.get_ylim()[1]],
           aspect="auto", alpha=0.4, zorder=1)

# Graficar datos experimentales y tericos
plt.plot(lambdas_exp, sin_thetas,"--o",color="grey")
ax1.plot(lambdas_exp, sin_thetas, "o", color="black", label="Datos experimentales", 
         markersize=10, linewidth=0, zorder=5, markerfacecolor='black', markeredgewidth=1.5, alpha=0.9)
ax1.plot(lambdas_teo, sin_thetas_teo, "s", color="firebrick", label="Valores tericos", 
         markersize=8, linewidth=0, zorder=4, alpha=0.9, markerfacecolor='firebrick', markeredgewidth=1.5)

# Conectar puntos experimentales y tericos con lneas
for i, (lam_exp, lam_teo, sin_t_exp, sin_t_teo) in enumerate(zip(lambdas_exp, lambdas_teo, sin_thetas, sin_thetas_teo)):
    ax1.plot([lam_exp, lam_teo], [sin_t_exp, sin_t_teo], 'k-', linewidth=1, alpha=0.5, zorder=2)

# Aadir lneas verticales para las longitudes de onda experimentales
for lam_exp, sin_t in zip(lambdas_exp, sin_thetas):
    ax1.axvline(lam_exp, color=wavelength_to_rgb(lam_exp), linewidth=2.5, alpha=0.8, linestyle='-', zorder=2)

# Aadir etiquetas para cada lnea espectral
line_labels = ['Violeta', 'Azul', 'Verde-azul', 'Verde', 'Amarillo', 'Naranja']
for i, (lam_exp, lam_teo, sin_t, label, error) in enumerate(zip(lambdas_exp, lambdas_teo, sin_thetas, line_labels, errors)):
    offset = 0.004 if i % 2 == 0 else -0.004
    ax1.annotate(f"{label}\n{lam_exp:.1f} nm\n={error:.1f}%", 
                 xy=(lam_exp, sin_t), 
                 xytext=(15, 25 if i % 2 == 0 else -35),
                 textcoords="offset points",
                 fontsize=12,
                 ha='center',
                 va='bottom' if i % 2 == 0 else 'top',
                 bbox=dict(boxstyle="round,pad=0.3", facecolor="w", alpha=0.9, edgecolor='gray'),
                 arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2", color='gray', lw=1))

# Mejorar el ttulo y etiquetas
ax1.set_title("Espectro del Mercurio: Relacin entre Longitud de Onda y seno del ngulo de Difraccin", 
              fontsize=22, pad=20)
ax1.set_xlabel(r" $\lambda$ [nm]", fontsize=18, labelpad=10)
ax1.set_ylabel(r"$\sin\theta$", fontsize=18, labelpad=10)

# Configurar lmites y ticks
ax1.set_xlim(395, 640)
ax1.set_ylim(0.22, 0.425)
ax1.xaxis.set_major_locator(MultipleLocator(50))
ax1.xaxis.set_minor_locator(MultipleLocator(10))
ax1.yaxis.set_major_locator(MultipleLocator(0.05))
ax1.yaxis.set_minor_locator(MultipleLocator(0.01))

# Aadir grid
ax1.grid(True, which='major', linestyle='--', linewidth=0.7, alpha=0.7)
ax1.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.5)

# Mejorar la leyenda
ax1.legend(loc='upper center', framealpha=0.9, shadow=True, fancybox=True)

# Segundo subplot: residuales
ax2 = fig.add_subplot(gs[1])
residuals = np.abs(lambdas_exp - lambdas_teo)*100/lambdas_teo

# Graficar residuales
#ax2.axhline(y=0, color='black', linestyle='-', alpha=0.5)
ax2.plot(lambdas_teo, residuals, '--o', color='black', markersize=8, alpha=0.8)

# Etiquetas y ttulo para residuales
ax2.set_xlabel(r"$\lambda$ [nm]", fontsize=18, labelpad=10)
ax2.set_ylabel(r"Error % con respecto al terico", fontsize=18, labelpad=15)
ax2.set_title("Errores con respecto a los valores tericos del Mercurio", fontsize=22, pad=15)

# Configurar grid para residuales
ax2.grid(True, which='major', linestyle='--', linewidth=0.7, alpha=0.7)
ax2.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.5)

# Ajustar diseo
plt.tight_layout()


plt.show()
```
