```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Función para cargar y limpiar datos
def cargar_datos_csv(nombre_csv):
    df = pd.read_csv(nombre_csv, sep=';', decimal=',', encoding='utf-8')

    col_angulo = [c for c in df.columns if "Ángulo" in c][0]
    col_intensidad = [c for c in df.columns if "Intensidad" in c][0]

    df_clean = df.dropna(subset=[col_angulo, col_intensidad], how='all')
    df_clean[col_angulo] = df_clean[col_angulo].fillna(method='ffill')
    df_clean = df_clean.dropna(subset=[col_intensidad])

    angulos = df_clean[col_angulo].values
    intensidades = df_clean[col_intensidad].values

    return angulos, intensidades

# Función para encontrar todos los picos
def encontrar_todos_los_picos(angulos, intensidades, umbral=50, distancia_minima=5):
    """
    Encuentra todos los picos en los datos
    umbral: intensidad mínima para considerar un pico
    distancia_minima: mínima distancia entre picos (en unidades de índice)
    """
    picos = []
    
    # Buscar máximos locales
    for i in range(1, len(intensidades)-1):
        # Condición para máximo local
        if (intensidades[i] > intensidades[i-1] and 
            intensidades[i] > intensidades[i+1] and 
            intensidades[i] > umbral):
            
            # Verificar que no esté muy cerca de otro pico
            if not picos or (angulos[i] - picos[-1][0]) > distancia_minima * (angulos[1] - angulos[0]):
                picos.append((angulos[i], intensidades[i]))
    
    return picos

# Cargar el primer archivo
ang1, inten1 = cargar_datos_csv("1.csv")

# Encontrar todos los picos
umbral_intensidad = 30  # Ajusta este valor según tus datos
picos_csv1 = encontrar_todos_los_picos(ang1, inten1, umbral=umbral_intensidad)

# Mostrar lista completa de picos
print("=" * 60)
print("LISTA COMPLETA DE PICOS - CSV 1")
print("=" * 60)
print(f"Número de picos encontrados: {len(picos_csv1)}")
print(f"Umbral de intensidad: {umbral_intensidad}%")
print("-" * 60)

for i, (angulo, intensidad) in enumerate(picos_csv1, 1):
    print(f"Pico {i:2d}: θ = {angulo:8.3f} rad | Intensidad = {intensidad:6.2f}% | {np.degrees(angulo):6.1f}°")

# Mostrar estadísticas
if picos_csv1:
    angulos_picos = [p[0] for p in picos_csv1]
    intensidades_picos = [p[1] for p in picos_csv1]
    
    print("-" * 60)
    print(f"Ángulo mínimo: {min(angulos_picos):.3f} rad")
    print(f"Ángulo máximo: {max(angulos_picos):.3f} rad")
    print(f"Rango angular: {max(angulos_picos) - min(angulos_picos):.3f} rad")
    print(f"Intensidad máxima: {max(intensidades_picos):.2f}%")
    print(f"Intensidad promedio: {np.mean(intensidades_picos):.2f}%")

# Configuración de estilo
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.size'] = 12
plt.rcParams['figure.dpi'] = 300

# Crear gráfico
fig, ax = plt.subplots(figsize=(14, 8))

# Graficar datos
ax.plot(ang1, inten1, "-", color="blue", linewidth=1.5, alpha=0.8, label="Datos CSV 1")

# Marcar todos los picos
for i, (angulo, intensidad) in enumerate(picos_csv1):
    ax.plot(angulo, intensidad, 'o', markersize=8, 
            color='red', markeredgecolor='white', markeredgewidth=1.5)
    
    # Añadir etiqueta numerada
    ax.text(angulo, intensidad + 2, f'{i+1}', 
            fontsize=9, ha='center', va='bottom', weight='bold',
            bbox=dict(boxstyle="circle,pad=0.2", facecolor="white", alpha=0.8))

# Configurar gráfico
ax.set_title("Todos los Picos Detectados - CSV 1", fontsize=16, pad=20)
ax.set_xlabel("Ángulo θ [rad]", fontsize=14, labelpad=10)
ax.set_ylabel("Intensidad luminosa (%)", fontsize=14, labelpad=10)
ax.grid(True, alpha=0.3)
ax.legend()

# Añadir tabla con los picos principales
if picos_csv1:
    # Tomar los 10 picos más intensos para la tabla
    picos_ordenados = sorted(picos_csv1, key=lambda x: x[1], reverse=True)[:10]
    
    tabla_texto = "Picos más intensos:\n"
    for i, (angulo, intensidad) in enumerate(picos_ordenados, 1):
        tabla_texto += f"{i}. θ={angulo:.2f} rad ({intensidad:.1f}%)\n"
    
    ax.text(0.02, 0.98, tabla_texto, transform=ax.transAxes, fontsize=10,
            verticalalignment='top', bbox=dict(boxstyle="round", facecolor="white", alpha=0.9))

plt.tight_layout()
plt.show()

# Versión alternativa: picos ordenados por intensidad
print("\n" + "=" * 60)
print("PICOS ORDENADOS POR INTENSIDAD (de mayor a menor)")
print("=" * 60)

picos_ordenados_intensidad = sorted(picos_csv1, key=lambda x: x[1], reverse=True)
for i, (angulo, intensidad) in enumerate(picos_ordenados_intensidad, 1):
    print(f"{i:2d}. θ = {angulo:8.3f} rad | {intensidad:6.2f}% | {np.degrees(angulo):6.1f}°")

# Versión alternativa: picos ordenados por ángulo
print("\n" + "=" * 60)
print("PICOS ORDENADOS POR ÁNGULO (de menor a mayor)")
print("=" * 60)

picos_ordenados_angulo = sorted(picos_csv1, key=lambda x: x[0])
for i, (angulo, intensidad) in enumerate(picos_ordenados_angulo, 1):
    print(f"{i:2d}. θ = {angulo:8.3f} rad | {intensidad:6.2f}% | {np.degrees(angulo):6.1f}°")

# Función para buscar picos en un rango específico
def buscar_picos_en_rango(picos, rango_angulo):
    """Busca picos dentro de un rango angular específico"""
    min_ang, max_ang = rango_angulo
    return [p for p in picos if min_ang <= p[0] <= max_ang]

# Ejemplo: buscar picos alrededor de 74° (1.29 rad)
print("\n" + "=" * 60)
print("PICOS CERCA DE 74° (1.29 rad)")
print("=" * 60)

picos_74 = buscar_picos_en_rango(picos_csv1, (1.20, 1.40))
for i, (angulo, intensidad) in enumerate(picos_74, 1):
    print(f"{i:2d}. θ = {angulo:8.3f} rad | {intensidad:6.2f}% | {np.degrees(angulo):6.1f}°")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Función para cargar y limpiar datos
def cargar_datos_csv(nombre_csv):
    df = pd.read_csv(nombre_csv, sep=';', decimal=',', encoding='utf-8')

    col_angulo = [c for c in df.columns if "Ángulo" in c][0]
    col_intensidad = [c for c in df.columns if "Intensidad" in c][0]

    df_clean = df.dropna(subset=[col_angulo, col_intensidad], how='all')
    df_clean[col_angulo] = df_clean[col_angulo].fillna(method='ffill')
    df_clean = df_clean.dropna(subset=[col_intensidad])

    angulos = df_clean[col_angulo].values
    intensidades = df_clean[col_intensidad].values

    return angulos, intensidades

# Función mejorada para encontrar TODOS los picos
def encontrar_todos_los_picos(angulos, intensidades, altura_minima=3, distancia_minima=10, prominencia=2):
    """
    Encuentra todos los picos en todo el rango angular
    altura_minima: intensidad mínima para considerar un pico (4% ahora)
    distancia_minima: mínima distancia entre picos (en unidades de índice)
    prominencia: qué tan sobresaliente debe ser el pico
    """
    # Usar find_peaks de scipy para mejor detección
    indices_picos, propiedades = find_peaks(
        intensidades, 
        height=altura_minima,  # ← AQUÍ se pone el 4%
        distance=distancia_minima,
        prominence=prominencia
    )
    
    picos = [(angulos[i], intensidades[i]) for i in indices_picos]
    return picos, propiedades

# Cargar el primer archivo
ang1, inten1 = cargar_datos_csv("1.csv")

# Encontrar TODOS los picos CON UMBRAL DE 4%
picos_csv1, propiedades = encontrar_todos_los_picos(ang1, inten1, altura_minima=4, distancia_minima=15, prominencia=2)

# Mostrar lista completa de picos
print("=" * 70)
print("LISTA COMPLETA DE TODOS LOS PICOS - CSV 1")
print("=" * 70)
print(f"Número total de picos encontrados: {len(picos_csv1)}")
print(f"Umbral de intensidad: 4%")
print(f"Rango angular: {ang1[0]:.1f} a {ang1[-1]:.1f} rad")
print("-" * 70)

# Agrupar picos por regiones
def agrupar_picos_por_region(picos, limites_regiones):
    """Agrupa picos en diferentes regiones angulares"""
    grupos = {}
    for lim_inf, lim_sup, nombre in limites_regiones:
        grupo = [p for p in picos if lim_inf <= p[0] <= lim_sup]
        grupos[nombre] = grupo
    return grupos

# Definir regiones (ajusta estos valores según tus datos)
limites_regiones = [
    (-135, -60, "Extremo izquierdo"),
    (-60, -20, "Izquierdo"),
    (-20, 20, "Región central"),
    (20, 60, "Derecho"), 
    (60, 135, "Extremo derecho")
]

picos_por_region = agrupar_picos_por_region(picos_csv1, limites_regiones)

# Mostrar picos por regiones
for region, picos_region in picos_por_region.items():
    print(f"\n{region.upper()} ({len(picos_region)} picos):")
    print("-" * 40)
    for i, (angulo, intensidad) in enumerate(picos_region, 1):
        print(f"  Pico {i}: θ = {angulo:7.2f} rad | {intensidad:5.1f}% | {np.degrees(angulo):6.1f}°")

# Mostrar todos los picos en orden angular
print("\n" + "=" * 70)
print("TODOS LOS PICOS ORDENADOS POR ÁNGULO (≥4% intensidad)")
print("=" * 70)

picos_ordenados = sorted(picos_csv1, key=lambda x: x[0])
for i, (angulo, intensidad) in enumerate(picos_ordenados, 1):
    region = next((nombre for lim_inf, lim_sup, nombre in limites_regiones if lim_inf <= angulo <= lim_sup), "Desconocida")
    print(f"{i:3d}. θ = {angulo:7.2f} rad | {intensidad:5.1f}% | {np.degrees(angulo):6.1f}° | {region}")

# Configuración de estilo
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.size'] = 10
plt.rcParams['figure.dpi'] = 300

# Crear gráfico completo
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))

# Gráfico 1: Vista completa con todos los picos
ax1.plot(ang1, inten1, "-", color="blue", linewidth=1.2, alpha=0.8, label="Datos CSV 1")

# Marcar todos los picos con colores por región
colores_regiones = ['red', 'orange', 'green', 'purple', 'brown']
for i, (lim_inf, lim_sup, nombre) in enumerate(limites_regiones):
    color = colores_regiones[i]
    picos_region = picos_por_region[nombre]
    
    for angulo, intensidad in picos_region:
        ax1.plot(angulo, intensidad, 'o', markersize=5, 
                color=color, markeredgecolor='white', markeredgewidth=0.8,
                label=nombre if angulo == picos_region[0][0] else "")
        
        # Etiqueta solo los picos más importantes (opcional)
        if intensidad > 20:  # Solo etiquetar picos más intensos
            ax1.text(angulo, intensidad + 2, f'{np.degrees(angulo):.0f}°', 
                    fontsize=7, ha='center', va='bottom', color=color,
                    bbox=dict(boxstyle="round,pad=0.1", facecolor="white", alpha=0.7))

ax1.set_title("Todos los Picos Detectados (≥4% intensidad) - CSV 1", fontsize=14, pad=20)
ax1.set_xlabel("Ángulo θ [rad]", fontsize=12, labelpad=10)
ax1.set_ylabel("Intensidad luminosa (%)", fontsize=12, labelpad=10)
ax1.grid(True, alpha=0.2)
ax1.legend(loc='upper right')
ax1.set_xlim(ang1[0], ang1[-1])

# Gráfico 2: Vista de cerca de regiones específicas
ax2.plot(ang1, inten1, "-", color="blue", linewidth=1.5, alpha=0.8)

# Marcar picos con números secuenciales
for i, (angulo, intensidad) in enumerate(picos_ordenados, 1):
    ax2.plot(angulo, intensidad, 'o', markersize=5, 
            color='red', markeredgecolor='white', markeredgewidth=0.8)
    
    # Etiquetar con número (solo si la intensidad es suficiente)
    if intensidad > 8:  # Etiquetar solo picos con >8% para no saturar
        ax2.text(angulo, intensidad + 1.5, str(i), fontsize=6, ha='center', va='bottom',
                bbox=dict(boxstyle="circle,pad=0.1", facecolor="white", alpha=0.8))

ax2.set_title("Todos los Picos Numerados (≥4% intensidad)", fontsize=14, pad=20)
ax2.set_xlabel("Ángulo θ [rad]", fontsize=12, labelpad=10)
ax2.set_ylabel("Intensidad luminosa (%)", fontsize=12, labelpad=10)
ax2.grid(True, alpha=0.2)
ax2.set_xlim(ang1[0], ang1[-1])

plt.tight_layout()
plt.show()

# Estadísticas de picos por rango de intensidad
print("\n" + "=" * 80)
print("ESTADÍSTICAS POR RANGO DE INTENSIDAD")
print("=" * 80)

rangos_intensidad = [
    (4, 10, "Débiles (4-10%)"),
    (10, 30, "Medios (10-30%)"),
    (30, 70, "Fuertes (30-70%)"),
    (70, 101, "Muy fuertes (70-100%)")
]

for lim_inf, lim_sup, categoria in rangos_intensidad:
    picos_categoria = [p for p in picos_csv1 if lim_inf <= p[1] < lim_sup]
    print(f"{categoria}: {len(picos_categoria)} picos")

# Mostrar los picos más intensos
print("\n" + "=" * 80)
print("10 PICOS MÁS INTENSOS")
print("=" * 80)

picos_mas_intensos = sorted(picos_csv1, key=lambda x: x[1], reverse=True)[:10]
for i, (angulo, intensidad) in enumerate(picos_mas_intensos, 1):
    print(f"{i:2d}. θ = {angulo:6.2f} rad | {intensidad:5.1f}% | {np.degrees(angulo):5.1f}°")

# Función para exportar datos
def exportar_picos_a_csv(picos, nombre_archivo):
    """Exporta la lista de picos a un archivo CSV"""
    datos = []
    for i, (angulo, intensidad) in enumerate(picos, 1):
        datos.append({
            'Número': i,
            'Ángulo_rad': angulo,
            'Ángulo_grados': np.degrees(angulo),
            'Intensidad_%': intensidad,
            'Región': next((nombre for lim_inf, lim_sup, nombre in limites_regiones if lim_inf <= angulo <= lim_sup), "Desconocida")
        })
    
    df = pd.DataFrame(datos)
    df.to_csv(nombre_archivo, index=False, encoding='utf-8')
    print(f"\nDatos exportados a {nombre_archivo}")

# Exportar resultados
exportar_picos_a_csv(picos_ordenados, "todos_los_picos_csv1_4porciento.csv")
```
