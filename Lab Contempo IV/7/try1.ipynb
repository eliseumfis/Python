```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import linregress
import pandas as pd

# ============================================================================
# DATOS EXPERIMENTALES
# ============================================================================
# Parte I: Ley del Inverso Cuadrado
distancias = np.array([2.4, 3.4, 4.4, 5.4, 6.4, 7.4, 8.4, 9.4, 10.4, 11.4])  # [cm]
rad_sinruido = np.array([3655, 1705, 1081, 770, 561, 407, 316, 288, 237, 208])
tiempo_preset = 30  # segundos

# Parte II: Absorcin de Partculas Beta
grosor = np.array([0.78, 0.4, 0.32, 0.21, 0.075, 0.01, 0.01016, 0.00178])  # [mg/cm]
rad_absorcion = np.array([20, 14, 14, 64, 655, 1190, 1640, 1658])
tiempo_preset_2 = 60  # segundos

# ============================================================================
# PARMETROS DEL CONTADOR GEIGER
# ============================================================================
tiempo_muerto = 5e-6  # [segundos] - aproximadamente 5 microsegundos
voltaje_GM = 900  # [voltios]

# ============================================================================
# PARTE I: LEY DEL INVERSO CUADRADO
# ============================================================================
print("="*70)
print("PARTE I: LEY DEL INVERSO CUADRADO")
print("="*70)

# Determinar radiacin de fondo (primera medicin sin fuente)
# Usaremos la primera medicin como referencia
fondo = 0  # En este caso, consideramos que ya est restado

# Correccin por tiempo muerto
def corregir_tiempo_muerto(cuentas_observadas, tiempo_muerto, tasa_conteo_observada):
    """Corrige las cuentas por tiempo muerto del contador Geiger"""
    # Correccin aproximada: 0.01% por cada 100 cpm
    tasa_conteo = (cuentas_observadas / tiempo_preset) * 60  # conversin a cpm
    factor_correccion = 1 + (0.0001 * tasa_conteo / 100)
    cuentas_corregidas = cuentas_observadas * factor_correccion
    return cuentas_corregidas, tasa_conteo

# Aplicar correccin a Parte I
cuentas_corregidas_1 = []
tasas_conteo_1 = []

for i, counts in enumerate(rad_sinruido):
    counts_corr, tasa = corregir_tiempo_muerto(counts, tiempo_muerto, 
                                                (counts/tiempo_preset)*60)
    cuentas_corregidas_1.append(counts_corr)
    tasas_conteo_1.append(tasa)

cuentas_corregidas_1 = np.array(cuentas_corregidas_1)
tasas_conteo_1 = np.array(tasas_conteo_1)

# Convertir distancias a metros
distancias_m = distancias / 100  # [m]

# Calcular 1/d
inverso_cuadrado = 1 / (distancias_m ** 2)

# Crear DataFrame para Parte I
df_parte1 = pd.DataFrame({
    'Distancia (cm)': distancias,
    'Distancia (m)': distancias_m,
    '1/d (1/m)': inverso_cuadrado,
    'Cuentas Observadas': rad_sinruido,
    'Tasa Conteo (cpm)': tasas_conteo_1,
    'Cuentas Corregidas': cuentas_corregidas_1
})

print("\nDatos Parte I - Ley del Inverso Cuadrado:")
print(df_parte1.to_string(index=False))

# Ajuste lineal: Cuentas vs 1/d
slope_1, intercept_1, r_value_1, p_value_1, std_err_1 = linregress(inverso_cuadrado, cuentas_corregidas_1)
r_squared_1 = r_value_1 ** 2

print(f"\n--- Anlisis del Ajuste Lineal (Cuentas vs 1/d) ---")
print(f"Pendiente (m): {slope_1:.2f}")
print(f"Intercepto (b): {intercept_1:.2f}")
print(f"R = {r_squared_1:.6f}")
print(f"Correlacin r = {r_value_1:.6f}")
print(f"Ecuacin: Cuentas = {slope_1:.2f}  (1/d) + {intercept_1:.2f}")

# Grfico Parte I
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Grfico 1: Cuentas vs Distancia
ax1 = axes[0]
ax1.scatter(distancias, rad_sinruido, color='red', s=80, label='Datos observados', zorder=3)
ax1.scatter(distancias, cuentas_corregidas_1, color='blue', s=60, 
            marker='^', label='Datos corregidos (t. muerto)', zorder=3)
ax1.set_xlabel('Distancia (cm)', fontsize=11, fontweight='bold')
ax1.set_ylabel('Cuentas', fontsize=11, fontweight='bold')
ax1.set_title('Parte I: Cuentas vs Distancia', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_xlim(1.5, 12)

# Grfico 2: Cuentas vs 1/d con lnea de tendencia
ax2 = axes[1]
ax2.scatter(inverso_cuadrado, cuentas_corregidas_1, color='green', s=80, 
            label='Datos experimentales', zorder=3)

# Lnea de tendencia
x_trend = np.linspace(0, max(inverso_cuadrado)*1.1, 100)
y_trend = slope_1 * x_trend + intercept_1
ax2.plot(x_trend, y_trend, 'r-', linewidth=2, label=f'Ajuste lineal (R={r_squared_1:.4f})')

ax2.set_xlabel('1/d (m)', fontsize=11, fontweight='bold')
ax2.set_ylabel('Cuentas', fontsize=11, fontweight='bold')
ax2.set_title('Parte I: Ley del Inverso Cuadrado', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.text(0.98, 0.02, f'y = {slope_1:.2f}x + {intercept_1:.2f}\nR = {r_squared_1:.6f}',
         transform=ax2.transAxes, verticalalignment='bottom', horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5), fontsize=10)

plt.tight_layout()
plt.savefig('parte_1_inverso_cuadrado.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# PARTE II: ABSORCIN DE PARTCULAS BETA
# ============================================================================
print("\n" + "="*70)
print("PARTE II: ABSORCIN DE PARTCULAS BETA")
print("="*70)

# Radiacin de fondo (primera medicin con mayor espesor)
fondo_2 = 0  # Ya consideramos que est restado

# Correccin por tiempo muerto (Parte II)
cuentas_corregidas_2 = []
cuentas_corregidas_fondo = []

for counts in rad_absorcion:
    counts_corr, _ = corregir_tiempo_muerto(counts, tiempo_muerto, 
                                             (counts/tiempo_preset_2)*60)
    cuentas_corregidas_2.append(counts_corr)

cuentas_corregidas_2 = np.array(cuentas_corregidas_2)

# Crear DataFrame para Parte II
df_parte2 = pd.DataFrame({
    'Espesor (mg/cm)': grosor,
    'Cuentas Observadas': rad_absorcion,
    'Cuentas Corregidas (t. muerto)': cuentas_corregidas_2,
    'Ln(Cuentas Corregidas)': np.log(cuentas_corregidas_2)
})

print("\nDatos Parte II - Absorcin de Partculas Beta:")
print(df_parte2.to_string(index=False))

# Anlisis de absorcin exponencial: I = I  e^(-x)
# Donde: I = intensidad (cuentas),  = coeficiente de absorcin, x = espesor
# Tomando logaritmo natural: ln(I) = ln(I) - x

# Ajuste logartmico
ln_cuentas = np.log(cuentas_corregidas_2)
slope_2, intercept_2, r_value_2, p_value_2, std_err_2 = linregress(grosor, ln_cuentas)
r_squared_2 = r_value_2 ** 2

# Coeficiente de absorcin
mu = -slope_2  # El coeficiente de absorcin

print(f"\n--- Anlisis de Absorcin Exponencial ---")
print(f"Ecuacin: ln(I) = ln(I) - x")
print(f"Intercepto ln(I): {intercept_2:.4f}")
print(f"I (intensidad inicial): {np.exp(intercept_2):.2f} cuentas")
print(f"Pendiente: {slope_2:.4f}")
print(f"Coeficiente de absorcin (): {mu:.4f} (mg/cm)")
print(f"R = {r_squared_2:.6f}")
print(f"Correlacin r = {r_value_2:.6f}")

# Rango extrapolado (interseccin con eje x cuando ln(I) = 0)
if slope_2 != 0:
    rango_extrap = -intercept_2 / slope_2
    print(f"Rango Extrapolado: {rango_extrap:.4f} mg/cm")

# Grfico Parte II
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Grfico 1: Cuentas vs Espesor (escala semilogartmica)
ax1 = axes[0]
ax1.scatter(grosor, rad_absorcion, color='purple', s=80, label='Datos observados', zorder=3)
ax1.scatter(grosor, cuentas_corregidas_2, color='orange', s=60, 
            marker='^', label='Corregidos (t. muerto)', zorder=3)
ax1.set_xlabel('Espesor (mg/cm)', fontsize=11, fontweight='bold')
ax1.set_ylabel('Cuentas', fontsize=11, fontweight='bold')
ax1.set_title('Parte II: Cuentas vs Espesor del Absorbente', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_yscale('log')

# Grfico 2: ln(Cuentas) vs Espesor con lnea de tendencia
ax2 = axes[1]
ax2.scatter(grosor, ln_cuentas, color='darkgreen', s=80, 
            label='Datos experimentales', zorder=3)

# Lnea de tendencia
x_trend_2 = np.linspace(0, max(grosor)*1.1, 100)
y_trend_2 = slope_2 * x_trend_2 + intercept_2
ax2.plot(x_trend_2, y_trend_2, 'r-', linewidth=2, label=f'Ajuste exponencial (R={r_squared_2:.4f})')
ax2.axhline(y=0, color='k', linestyle='--', linewidth=0.8, alpha=0.5)
if slope_2 != 0:
    ax2.axvline(x=rango_extrap, color='b', linestyle='--', linewidth=1.5, 
                label=f'Rango extrapolado = {rango_extrap:.4f} mg/cm')

ax2.set_xlabel('Espesor (mg/cm)', fontsize=11, fontweight='bold')
ax2.set_ylabel('ln(Cuentas)', fontsize=11, fontweight='bold')
ax2.set_title('Parte II: Absorcin Exponencial de Partculas Beta', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.text(0.98, 0.02, f'y = {slope_2:.4f}x + {intercept_2:.4f}\n = {mu:.4f} (mg/cm)\nR = {r_squared_2:.6f}',
         transform=ax2.transAxes, verticalalignment='bottom', horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5), fontsize=10)

plt.tight_layout()
plt.savefig('parte_2_absorcion_beta.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# CONCLUSIONES
# ============================================================================
print("\n" + "="*70)
print("CONCLUSIONES")
print("="*70)

print("\n--- PARTE I: LEY DEL INVERSO CUADRADO ---")
if r_squared_1 > 0.95:
    print(f" Los datos OBEDECEN muy bien la ley del inverso cuadrado (R = {r_squared_1:.6f})")
elif r_squared_1 > 0.90:
    print(f" Los datos OBEDECEN bien la ley del inverso cuadrado (R = {r_squared_1:.6f})")
else:
    print(f" Los datos muestran una correlacin moderada (R = {r_squared_1:.6f})")

print(f"\nLa relacin es lineal: Cuentas = {slope_1:.2f}  (1/d) + {intercept_1:.2f}")
print(f"Esto confirma que la intensidad es inversamente proporcional al cuadrado de la distancia.")

print("\n--- PARTE II: ABSORCIN DE PARTCULAS BETA ---")
if r_squared_2 > 0.95:
    print(f" El ajuste exponencial es EXCELENTE (R = {r_squared_2:.6f})")
elif r_squared_2 > 0.90:
    print(f" El ajuste exponencial es MUY BUENO (R = {r_squared_2:.6f})")
else:
    print(f" El ajuste exponencial es moderado (R = {r_squared_2:.6f})")

print(f"\nLa atenuacin sigue la forma: I = {np.exp(intercept_2):.2f}  e^(-{mu:.4f}x)")
print(f"Coeficiente de absorcin () = {mu:.4f} (mg/cm)")
if slope_2 != 0:
    print(f"Rango extrapolado = {rango_extrap:.4f} mg/cm")
print(f"\nEsto demuestra que la radiacin beta se atena exponencialmente con el espesor del material absorbente.")
print("="*70)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import pandas as pd

# ============================================================================
# DATOS EXPERIMENTALES (YA CORREGIDOS)
# ============================================================================
# Parte I: Ley del Inverso Cuadrado
distancias = np.array([2.4, 3.4, 4.4, 5.4, 6.4, 7.4, 8.4, 9.4, 10.4, 11.4])  # [cm]
cuentas_1 = np.array([3655, 1705, 1081, 770, 561, 407, 316, 288, 237, 208])

# Parte II: Absorcin de Partculas Beta
grosor = np.array([0.78, 0.4, 0.32, 0.21, 0.075, 0.01, 0.01016, 0.00178])  # [mg/cm]
cuentas_2 = np.array([20, 14, 14, 64, 655, 1190, 1640, 1658])

# ============================================================================
# PARTE I: LEY DEL INVERSO CUADRADO
# ============================================================================
print("="*75)
print(" PARTE I: LEY DEL INVERSO CUADRADO DE LA RADIACIN")
print("="*75)

# Convertir distancias a metros
distancias_m = distancias / 100  # [m]

# Calcular 1/d
inverso_cuadrado = 1 / (distancias_m ** 2)

# Crear DataFrame para Parte I
df_parte1 = pd.DataFrame({
    'Distancia (cm)': distancias,
    'Distancia (m)': distancias_m,
    '1/d (m)': inverso_cuadrado,
    'Cuentas': cuentas_1
})

print("\nTABLA DE DATOS - PARTE I:")
print(df_parte1.to_string(index=False))

# Ajuste lineal: Cuentas vs 1/d
slope_1, intercept_1, r_value_1, p_value_1, std_err_1 = linregress(inverso_cuadrado, cuentas_1)
r_squared_1 = r_value_1 ** 2

print(f"\n--- ANLISIS DE REGRESIN LINEAL ---")
print(f"Pendiente (m): {slope_1:.4f}")
print(f"Intercepto (b): {intercept_1:.4f}")
print(f"Ecuacin: y = {slope_1:.4f}x + {intercept_1:.4f}")
print(f"Coeficiente de correlacin (R): {r_value_1:.6f}")
print(f"Coeficiente de determinacin (R): {r_squared_1:.6f}")

# Grfico Parte I
fig = plt.figure(figsize=(13, 5))

# Grfico 1: Cuentas vs Distancia
ax1 = plt.subplot(1, 2, 1)
ax1.scatter(distancias, cuentas_1, s=100, color='#E74C3C', edgecolors='black', linewidth=1.5, zorder=3)
ax1.set_xlabel('Distancia (cm)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Cuentas', fontsize=12, fontweight='bold')
ax1.set_title('Parte I: Cuentas vs Distancia', fontsize=13, fontweight='bold')
ax1.grid(True, alpha=0.4, linestyle='--')
ax1.set_xlim(1.5, 12)

# Grfico 2: Cuentas vs 1/d con lnea de tendencia
ax2 = plt.subplot(1, 2, 2)
ax2.scatter(inverso_cuadrado, cuentas_1, s=100, color='#3498DB', edgecolors='black', linewidth=1.5, zorder=3, label='Datos experimentales')

# Lnea de tendencia
x_trend = np.linspace(0, max(inverso_cuadrado)*1.05, 100)
y_trend = slope_1 * x_trend + intercept_1
ax2.plot(x_trend, y_trend, 'r-', linewidth=2.5, label='Ajuste lineal', zorder=2)

ax2.set_xlabel('1/d (m)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Cuentas', fontsize=12, fontweight='bold')
ax2.set_title('Parte I: Ley del Inverso Cuadrado', fontsize=13, fontweight='bold')
ax2.grid(True, alpha=0.4, linestyle='--')
ax2.legend(fontsize=11, loc='upper left')

# Caja de texto con ecuacin y R
textstr = f'y = {slope_1:.2f}x + {intercept_1:.2f}\nR = {r_squared_1:.6f}'
props = dict(boxstyle='round', facecolor='#F9E79F', alpha=0.9, edgecolor='black', linewidth=1.5)
ax2.text(0.98, 0.05, textstr, transform=ax2.transAxes, fontsize=11, fontweight='bold',
         verticalalignment='bottom', horizontalalignment='right', bbox=props)

plt.tight_layout()
plt.savefig('parte_1_inverso_cuadrado.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# PARTE II: ABSORCIN DE PARTCULAS BETA
# ============================================================================
print("\n" + "="*75)
print(" PARTE II: ABSORCIN DE PARTCULAS BETA")
print("="*75)

# Crear DataFrame para Parte II
df_parte2 = pd.DataFrame({
    'Espesor (mg/cm)': grosor,
    'Cuentas': cuentas_2,
    'ln(Cuentas)': np.log(cuentas_2)
})

print("\nTABLA DE DATOS - PARTE II:")
print(df_parte2.to_string(index=False))

# Anlisis de absorcin exponencial: I = I  e^(-x)
# ln(I) = ln(I) - x (regresin lineal)
ln_cuentas = np.log(cuentas_2)
slope_2, intercept_2, r_value_2, p_value_2, std_err_2 = linregress(grosor, ln_cuentas)
r_squared_2 = r_value_2 ** 2

# Coeficiente de absorcin y parmetros
mu = -slope_2
I0 = np.exp(intercept_2)

print(f"\n--- ANLISIS DE ABSORCIN EXPONENCIAL ---")
print(f"Ecuacin terica: I = I  e^(-x)")
print(f"Forma linealizada: ln(I) = ln(I) - x")
print(f"\nln(I): {intercept_2:.6f}")
print(f"I (intensidad inicial): {I0:.2f} cuentas")
print(f"Pendiente: {slope_2:.6f}")
print(f"Coeficiente de absorcin (): {mu:.6f} (mg/cm)")
print(f"Coeficiente de correlacin (R): {r_value_2:.6f}")
print(f"Coeficiente de determinacin (R): {r_squared_2:.6f}")
print(f"Ecuacin: ln(I) = {intercept_2:.6f} + ({slope_2:.6f})x")

# Rango extrapolado
if slope_2 != 0:
    rango_extrap = -intercept_2 / slope_2
    print(f"Rango extrapolado (I=1): {rango_extrap:.6f} mg/cm")

# Grfico Parte II
fig = plt.figure(figsize=(13, 5))

# Grfico 1: Cuentas vs Espesor (escala semilogartmica)
ax1 = plt.subplot(1, 2, 1)
ax1.scatter(grosor, cuentas_2, s=100, color='#27AE60', edgecolors='black', linewidth=1.5, zorder=3)
ax1.set_xlabel('Espesor del absorbente (mg/cm)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Cuentas', fontsize=12, fontweight='bold')
ax1.set_title('Parte II: Atenuacin de Radiacin Beta', fontsize=13, fontweight='bold')
ax1.set_yscale('log')
ax1.grid(True, alpha=0.4, linestyle='--', which='both')

# Grfico 2: ln(Cuentas) vs Espesor con lnea de tendencia
ax2 = plt.subplot(1, 2, 2)
ax2.scatter(grosor, ln_cuentas, s=100, color='#8E44AD', edgecolors='black', linewidth=1.5, zorder=3, label='Datos experimentales')

# Lnea de tendencia
x_trend_2 = np.linspace(0, max(grosor)*1.05, 100)
y_trend_2 = slope_2 * x_trend_2 + intercept_2
ax2.plot(x_trend_2, y_trend_2, 'r-', linewidth=2.5, label='Ajuste exponencial', zorder=2)

# Lnea y=0 para rango extrapolado
ax2.axhline(y=0, color='gray', linestyle='--', linewidth=1, alpha=0.7)
if slope_2 != 0:
    ax2.axvline(x=rango_extrap, color='green', linestyle='--', linewidth=2, alpha=0.7, label=f'Rango extrapolado = {rango_extrap:.4f}')

ax2.set_xlabel('Espesor del absorbente (mg/cm)', fontsize=12, fontweight='bold')
ax2.set_ylabel('ln(Cuentas)', fontsize=12, fontweight='bold')
ax2.set_title('Parte II: Absorcin Exponencial', fontsize=13, fontweight='bold')
ax2.grid(True, alpha=0.4, linestyle='--')
ax2.legend(fontsize=11, loc='upper right')

# Caja de texto con ecuacin y R
textstr = f'ln(I) = {intercept_2:.4f} + ({slope_2:.6f})x\n = {mu:.6f} (mg/cm)\nR = {r_squared_2:.6f}'
props = dict(boxstyle='round', facecolor='#F9E79F', alpha=0.9, edgecolor='black', linewidth=1.5)
ax2.text(0.98, 0.05, textstr, transform=ax2.transAxes, fontsize=10, fontweight='bold',
         verticalalignment='bottom', horizontalalignment='right', bbox=props)

plt.tight_layout()
plt.savefig('parte_2_absorcion_beta.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# CONCLUSIONES
# ============================================================================
print("\n" + "="*75)
print(" CONCLUSIONES")
print("="*75)

print("\n" + ""*75)
print("PARTE I: LEY DEL INVERSO CUADRADO")
print(""*75)

if r_squared_1 > 0.99:
    print(f" EXCELENTE: Los datos obedecen perfectamente la ley del inverso cuadrado")
elif r_squared_1 > 0.95:
    print(f" MUY BUENO: Los datos obedecen muy bien la ley del inverso cuadrado")
elif r_squared_1 > 0.90:
    print(f" BUENO: Los datos obedecen bien la ley del inverso cuadrado")
else:
    print(f" MODERADO: Los datos muestran correlacin moderada con la ley del inverso cuadrado")

print(f"\nR = {r_squared_1:.6f} indica que el {r_squared_1*100:.4f}% de la variacin en las cuentas")
print(f"se explica por la relacin inverso cuadrado.")

print(f"\nLa relacin lineal obtenida es:")
print(f"   Cuentas = {slope_1:.4f}  (1/d) + {intercept_1:.4f}")

print(f"\nInterpretacin:")
print(f" La pendiente ({slope_1:.4f}) representa la intensidad caracterstica de la fuente")
print(f" El intercepto ({intercept_1:.4f}) representa radiacin de fondo residual")
print(f" La alta correlacin (r = {r_value_1:.6f}) confirma la ley del inverso cuadrado")
print(f"\nCONCLUSIN: La intensidad de radiacin es inversamente proporcional al cuadrado")
print(f"de la distancia, verificando I  1/d")

print("\n" + ""*75)
print("PARTE II: ABSORCIN DE PARTCULAS BETA")
print(""*75)

if r_squared_2 > 0.99:
    print(f" EXCELENTE: El ajuste exponencial es prcticamente perfecto")
elif r_squared_2 > 0.95:
    print(f" MUY BUENO: El ajuste exponencial es muy bueno")
elif r_squared_2 > 0.90:
    print(f" BUENO: El ajuste exponencial es bueno")
else:
    print(f" MODERADO: El ajuste exponencial muestra correlacin moderada")

print(f"\nR = {r_squared_2:.6f} indica que el {r_squared_2*100:.4f}% de la variacin en ln(Cuentas)")
print(f"se explica por la ley de absorcin exponencial.")

print(f"\nLa ecuacin de absorcin obtenida es:")
print(f"   I(x) = {I0:.2f}  e^(-{mu:.6f}x)")
print(f"\nForma linealizada:")
print(f"   ln(I) = {intercept_2:.6f} - {mu:.6f}x")

print(f"\nParmetros fsicos:")
print(f" I (intensidad inicial): {I0:.2f} cuentas")
print(f"  (coeficiente de absorcin): {mu:.6f} (mg/cm)")
if slope_2 != 0:
    print(f" Rango extrapolado: {rango_extrap:.6f} mg/cm")
    print(f"  (espesor necesario para atenuar la intensidad a 1 cuenta)")

print(f"\nInterpretacin:")
print(f" La radiacin beta disminuye exponencialmente con el espesor del absorbente")
print(f" El coeficiente  depende de la energa de las partculas beta y del material absorbente")
print(f" La alta correlacin (r = {r_value_2:.6f}) confirma el comportamiento exponencial")

print(f"\nCONCLUSIN: Las partculas beta experimentan atenuacin exponencial al pasar")
print(f"a travs de materia, siguiendo la ley I = Ie^(-x), siendo interceptadas")
print(f"por colisiones con el material absorbente.")

print("\n" + "="*75)
```
