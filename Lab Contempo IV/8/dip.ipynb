```python
# Encontrar mximos
max_peaks, _ = find_peaks(I, height=1000, distance=10)
max_voltages = V[max_peaks]
max_currents = I[max_peaks]

# Encontrar mnimos
min_peaks, _ = find_peaks(-I, height=-2000, distance=10)
min_voltages = V[min_peaks]
min_currents = I[min_peaks]

print("MXIMOS RELATIVOS:")
print("Voltaje (V)\tCorriente (nA)")
for i in range(len(max_peaks)):
    print(f"{max_voltages[i]:.1f}\t\t{max_currents[i]}")

print("\nMNIMOS RELATIVOS:")
print("Voltaje (V)\tCorriente (nA)")
for i in range(len(min_peaks)):
    print(f"{min_voltages[i]:.1f}\t\t{min_currents[i]}")
```

```python
# Diferencias entre mximos consecutivos
delta_V_max = np.diff(max_voltages)
delta_V_min = np.diff(min_voltages)

print("DIFERENCIAS ENTRE MXIMOS:")
for i in range(len(delta_V_max)):
    print(f"Vmax{i+2} - Vmax{i+1} = {delta_V_max[i]:.2f} V")

print(f"\nPromedio V_max = {np.mean(delta_V_max):.2f} V")
print(f"Desviacin estndar V_max = {np.std(delta_V_max):.2f} V")

print("\nDIFERENCIAS ENTRE MNIMOS:")
for i in range(len(delta_V_min)):
    print(f"Vmin{i+2} - Vmin{i+1} = {delta_V_min[i]:.2f} V")

print(f"\nPromedio V_min = {np.mean(delta_V_min):.2f} V")
print(f"Desviacin estndar V_min = {np.std(delta_V_min):.2f} V")
```
