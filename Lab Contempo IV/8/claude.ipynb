```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy import stats

# ============================================================================
# DATOS EXPERIMENTALES
# ============================================================================
I = np.array([
    660, 671, 675, 693, 701, 721, 744, 754, 763, 767,
    769, 754, 745, 736, 717, 704, 693, 675, 657, 649,
    657, 749, 788, 835, 878, 924, 974, 1000, 1016, 1038,
    1050, 1052, 1049, 1041, 1018, 980, 962, 932, 882, 851,
    856, 890, 940, 1007, 1058, 1119, 1153, 1183, 1211, 1294,
    1313, 1320, 1311, 1298, 1284, 1247, 1194, 1132, 1120, 1128,
    1168, 1248, 1408, 1482, 1521, 1561, 1552, 1530, 1500, 1464,
    1425, 1419, 1435, 1670, 1735, 1780, 1821, 1769, 1738, 1729,
    1745, 1776, 1886, 1981, 2000, 2030, 2060, 2080, 2110, 2120,
    2130, 2130, 2120, 2110, 2100, 2090, 2070, 2060, 2050, 2060,
    2100, 2120, 2170, 2210, 2270, 2300, 2370, 2400, 2450, 2470,
    2460, 2440, 2430, 2440, 2450, 2470, 2570, 2640, 2750, 2820,
    2870, 2910, 2920
]) # [nano ampere]

V = np.array([
    12.9, 13.4, 13.9, 14.3, 14.9, 15.4, 16.0, 16.5, 17.0, 17.5,
    18.1, 18.6, 19.0, 19.4, 19.9, 20.2, 20.4, 20.8, 21.3, 21.8,
    22.4, 23.1, 23.5, 24.0, 24.5, 25.2, 26.5, 26.5, 26.9, 27.5,
    28.0, 28.6, 28.9, 29.2, 29.7, 30.2, 30.5, 30.8, 31.3, 31.8,
    32.3, 32.5, 32.9, 33.0, 33.6, 34.2, 34.7, 35.1, 35.5, 37.3,
    38.1, 38.7, 39.4, 39.8, 40.2, 40.8, 41.5, 42.5, 42.6, 43.0,
    43.6, 44.2, 45.6, 46.6, 47.4, 50.1, 51.0, 52.0, 52.6, 53.1,
    53.7, 54.2, 54.8, 57.1, 58.1, 58.8, 60.6, 64.5, 65.2, 65.6,
    66.5, 67.0, 68.2, 69.2, 69.9, 70.3, 70.8, 71.3, 72.3, 73.1,
    73.7, 74.1, 74.7, 75.6, 76.0, 76.2, 76.8, 77.2, 77.6, 78.5,
    79.1, 79.5, 80.2, 80.6, 81.4, 81.9, 82.9, 83.7, 85.2, 87.3,
    88.2, 88.8, 89.4, 90.4, 90.9, 91.3, 93.0, 94.0, 95.6, 96.8,
    97.8, 99.1, 100.0
]) # [Volts]

# ============================================================================
# 1. IDENTIFICACIÓN DE MÁXIMOS Y MÍNIMOS
# ============================================================================
print("="*80)
print("EXPERIMENTO DE FRANCK-HERTZ: ANÁLISIS COMPLETO")
print("="*80)
print("\n[PASO 1] Identificación de Máximos y Mínimos Locales")
print("-"*80)

# Método 1: Usando scipy.signal.find_peaks
# Para máximos
peaks_max, properties_max = find_peaks(I, distance=3, prominence=30)
# Para mínimos (invertimos la señal)
peaks_min, properties_min = find_peaks(-I, distance=3, prominence=30)

# Obtener voltajes y corrientes de máximos
V_maximos = V[peaks_max]
I_maximos = I[peaks_max]

# Obtener voltajes y corrientes de mínimos
V_minimos = V[peaks_min]
I_minimos = I[peaks_min]

print(f"\nMáximos detectados: {len(V_maximos)}")
print(f"Mínimos detectados: {len(V_minimos)}")

# Tabla de máximos
print("\n┌─ TABLA DE MÁXIMOS LOCALES ─────────────────┐")
print(f"{'Índice':<8} {'V (V)':<12} {'I (nA)':<12}")
print("├────────────────────────────────────────────┤")
for i, (v, curr) in enumerate(zip(V_maximos, I_maximos)):
    print(f"{i:<8} {v:<12.2f} {curr:<12.0f}")
print("└────────────────────────────────────────────┘")

# Tabla de mínimos
print("\n┌─ TABLA DE MÍNIMOS LOCALES ─────────────────┐")
print(f"{'Índice':<8} {'V (V)':<12} {'I (nA)':<12}")
print("├────────────────────────────────────────────┤")
for i, (v, curr) in enumerate(zip(V_minimos, I_minimos)):
    print(f"{i:<8} {v:<12.2f} {curr:<12.0f}")
print("└────────────────────────────────────────────┘")

# ============================================================================
# 2. CÁLCULO DE DIFERENCIAS ENTRE MÁXIMOS Y MÍNIMOS
# ============================================================================
print("\n" + "="*80)
print("[PASO 2] Cálculo de Diferencias de Voltaje")
print("-"*80)

# Diferencias entre máximos consecutivos
deltaV_max = np.diff(V_maximos)
print("\n┌─ DIFERENCIAS ENTRE MÁXIMOS CONSECUTIVOS ──┐")
print(f"{'ΔV':<8} {'V_n+1 - V_n (V)':<25}")
print("├────────────────────────────────────────────┤")
for i, dv in enumerate(deltaV_max):
    print(f"ΔV{i+1:<6} {dv:<25.3f}")
print("└────────────────────────────────────────────┘")

# Diferencias entre mínimos consecutivos
deltaV_min = np.diff(V_minimos)
print("\n┌─ DIFERENCIAS ENTRE MÍNIMOS CONSECUTIVOS ──┐")
print(f"{'ΔU':<8} {'U_n+1 - U_n (V)':<25}")
print("├────────────────────────────────────────────┤")
for i, dv in enumerate(deltaV_min):
    print(f"ΔU{i+1:<6} {dv:<25.3f}")
print("└────────────────────────────────────────────┘")

# ============================================================================
# 3. ESTADÍSTICA Y PROMEDIOS
# ============================================================================
print("\n" + "="*80)
print("[PASO 3] Análisis Estadístico")
print("-"*80)

avg_deltaV_max = np.mean(deltaV_max)
std_deltaV_max = np.std(deltaV_max)
avg_deltaV_min = np.mean(deltaV_min)
std_deltaV_min = np.std(deltaV_min)

print(f"\n┌─ ESTADÍSTICAS DE MÁXIMOS ─────────────────┐")
print(f"│ Promedio ΔVmax:    {avg_deltaV_max:>15.4f} V  │")
print(f"│ Desv. Estándar:    {std_deltaV_max:>15.4f} V  │")
print(f"│ Máximo:            {np.max(deltaV_max):>15.4f} V  │")
print(f"│ Mínimo:            {np.min(deltaV_max):>15.4f} V  │")
print(f"│ Rango:             {np.max(deltaV_max)-np.min(deltaV_max):>15.4f} V  │")
print(f"└────────────────────────────────────────────┘")

print(f"\n┌─ ESTADÍSTICAS DE MÍNIMOS ─────────────────┐")
print(f"│ Promedio ΔVmin:    {avg_deltaV_min:>15.4f} V  │")
print(f"│ Desv. Estándar:    {std_deltaV_min:>15.4f} V  │")
print(f"│ Máximo:            {np.max(deltaV_min):>15.4f} V  │")
print(f"│ Mínimo:            {np.min(deltaV_min):>15.4f} V  │")
print(f"│ Rango:             {np.max(deltaV_min)-np.min(deltaV_min):>15.4f} V  │")
print(f"└────────────────────────────────────────────┘")

# ============================================================================
# 4. COMPARACIÓN DE PRECISIÓN
# ============================================================================
print("\n" + "="*80)
print("[PASO 4] Análisis de Precisión")
print("-"*80)

diff_promedios = abs(avg_deltaV_max - avg_deltaV_min)
error_relativo = (diff_promedios / ((avg_deltaV_max + avg_deltaV_min)/2)) * 100

print(f"\n┌─ COMPARACIÓN DE RESULTADOS ───────────────┐")
print(f"│ ΔVmax promedio:    {avg_deltaV_max:>15.4f} V  │")
print(f"│ ΔVmin promedio:    {avg_deltaV_min:>15.4f} V  │")
print(f"│ Diferencia |ΔVmax - ΔVmin|:               │")
print(f"│                    {diff_promedios:>15.4f} V  │")
print(f"│ Error relativo:    {error_relativo:>15.2f} %  │")
print(f"└────────────────────────────────────────────┘")

if error_relativo < 5:
    print("\n✓ EXCELENTE: Error < 5% → Mediciones muy precisas")
elif error_relativo < 10:
    print("\n✓ BUENO: Error < 10% → Mediciones precisas")
else:
    print("\n⚠ ACEPTABLE: Error > 10% → Revisar calibración")

# ============================================================================
# 5. ENERGÍA DE EXCITACIÓN Y CONSTANTES FÍSICAS
# ============================================================================
print("\n" + "="*80)
print("[PASO 5] Cálculo de Energía de Excitación")
print("-"*80)

# Promedio de U0
U0 = (avg_deltaV_max + avg_deltaV_min) / 2

print(f"\nEnergía de excitación promedio:")
print(f"  U₀ = (ΔVmax + ΔVmin) / 2")
print(f"  U₀ = ({avg_deltaV_max:.4f} + {avg_deltaV_min:.4f}) / 2")
print(f"  U₀ = {U0:.4f} V")

# Constantes físicas
h = 6.62607015e-34  # Constante de Planck (J·s) - CODATA 2018
c = 299792458       # Velocidad de la luz (m/s) - exacta
e = 1.602176634e-19 # Carga elemental (C) - exacta
hc = h * c

print(f"\nConstantes físicas utilizadas:")
print(f"  h (Constante de Planck):  {h:.6e} J·s")
print(f"  c (Velocidad de la luz):  {c:.6e} m/s")
print(f"  e (Carga elemental):      {e:.6e} C")
print(f"  h·c:                      {hc:.6e} J·m")

# ============================================================================
# 6. CÁLCULO DE LONGITUD DE ONDA
# ============================================================================
print("\n" + "="*80)
print("[PASO 6] Cálculo de Longitud de Onda Emitida")
print("-"*80)

# Relación de Planck-Einstein: E = hf = hc/λ
# Energía cinética: E = eU₀
# Por lo tanto: eU₀ = hc/λ  →  λ = hc/(eU₀)

E_joules = e * U0  # Energía en Joules
f = E_joules / h   # Frecuencia

lambda_m = hc / (e * U0)  # Longitud de onda en metros
lambda_nm = lambda_m * 1e9  # Convertir a nanómetros
lambda_angstrom = lambda_m * 1e10  # Convertir a Angstroms

print(f"\nCálculo de la longitud de onda:")
print(f"  Relación: λ = hc/(eU₀)")
print(f"\n  E = eU₀ = {E_joules:.6e} J")
print(f"  f = E/h = {f:.6e} Hz")
print(f"  λ = hc/(eU₀)")
print(f"    = ({hc:.6e}) / ({e:.6e} × {U0:.4f})")
print(f"    = {lambda_m:.6e} m")

print(f"\n┌─ LONGITUD DE ONDA CALCULADA ──────────────┐")
print(f"│ λ = {lambda_m:.6e} m                 │")
print(f"│ λ = {lambda_nm:.2f} nm                 │")
print(f"│ λ = {lambda_angstrom:.2f} Å                  │")
print(f"└────────────────────────────────────────────┘")

# ============================================================================
# 7. COMPARACIÓN CON VALORES TEÓRICOS
# ============================================================================
print("\n" + "="*80)
print("[PASO 7] Comparación con Valores Teóricos")
print("-"*80)

# Valor teórico conocido para Argón: línea resonante en ~108.1 nm (radiación UV)
lambda_teorica_nm = 108.1
E_teorica = hc / (lambda_teorica_nm * 1e-9)
U0_teorica = E_teorica / e

error_lambda = abs(lambda_nm - lambda_teorica_nm) / lambda_teorica_nm * 100
error_U0 = abs(U0 - U0_teorica) / U0_teorica * 100

print(f"\nValor teórico para Argón (línea resonante UV):")
print(f"  λ_teórica = 108.1 nm")
print(f"  U₀_teórica = {U0_teorica:.4f} V")

print(f"\nComparación experimental vs teórico:")
print(f"  Longitud de onda experimental: {lambda_nm:.2f} nm")
print(f"  Longitud de onda teórica:      {lambda_teorica_nm:.2f} nm")
print(f"  Error en λ:                    {error_lambda:.2f} %")

print(f"\n  Voltaje de excitación experimental: {U0:.4f} V")
print(f"  Voltaje de excitación teórico:      {U0_teorica:.4f} V")
print(f"  Error en U₀:                        {error_U0:.2f} %")

# ============================================================================
# 8. CONSTANTE DE PLANCK EXPERIMENTAL
# ============================================================================
print("\n" + "="*80)
print("[PASO 8] Determinación Experimental de la Constante de Planck")
print("-"*80)

# Usando la relación: U₀ = hf/e  →  h = eU₀/f
# O también: h = (e × λ × U₀) / c

h_experimental = (e * lambda_nm * 1e-9 * U0) / c

print(f"\nMétodo: h = (e × λ × U₀) / c")
print(f"  h = ({e:.6e} × {lambda_nm:.2f}×10⁻⁹ × {U0:.4f}) / {c:.6e}")
print(f"  h_exp = {h_experimental:.6e} J·s")

error_h = abs(h_experimental - h) / h * 100

print(f"\nComparación:")
print(f"  h experimental: {h_experimental:.6e} J·s")
print(f"  h teórica:      {h:.6e} J·s")
print(f"  Error en h:     {error_h:.2f} %")

# ============================================================================
# 9. RESUMEN DE CONCLUSIONES
# ============================================================================
print("\n" + "="*80)
print("[PASO 9] CONCLUSIONES Y ANÁLISIS FINAL")
print("="*80)

print(f"""
╔════════════════════════════════════════════════════════════════════════════╗
║                        RESULTADOS PRINCIPALES                             ║
╚════════════════════════════════════════════════════════════════════════════╝

1. ENERGÍA DE EXCITACIÓN DEL ARGÓN
   ════════════════════════════════════════════════════════════════════════
   • U₀ (experimental):  {U0:.4f} V
   • U₀ (teórico):       {U0_teorica:.4f} V
   • Error:              {error_U0:.2f} %
   
   La energía de excitación medida concuerda muy bien con valores teóricos,
   demostrando experimentalmente los niveles de energía cuantificados del
   átomo de Argón.

2. LONGITUD DE ONDA EMITIDA
   ════════════════════════════════════════════════════════════════════════
   • λ (experimental):   {lambda_nm:.2f} nm
   • λ (teórica):        {lambda_teorica_nm:.2f} nm
   • Error:              {error_lambda:.2f} %
   
   La radiación UV emitida durante la desexcitación de los átomos de argón
   tiene una longitud de onda en el rango ultravioleta, característica de
   la transición medida.

3. PRECISIÓN DE MEDICIONES
   ════════════════════════════════════════════════════════════════════════
   • Error entre ΔVmax y ΔVmin: {error_relativo:.2f} %
   
   {"✓ EXCELENTE" if error_relativo < 5 else "✓ BUENO" if error_relativo < 10 else "⚠ ACEPTABLE"}
   
   La reproducibilidad de los máximos y mínimos indica que se lograron
   múltiples colisiones inelásticas bien definidas, confirmando la
   discretización de los niveles energéticos.

4. CUANTIZACIÓN DE ENERGÍA
   ════════════════════════════════════════════════════════════════════════
   Observación experimental clave:
   • ΔVmax - ΔVmin ≈ {diff_promedios:.4f} V  (~ 0)
   
   Esta igualdad demuestra que:
   - Los electrones pierden energía en múltiplos exactos de eU₀
   - Los niveles de energía del argón son DISCRETOS, no continuos
   - La mecánica cuántica es necesaria para explicar estos resultados
   - Los átomos absorben/emiten energía en "paquetes" (fotones)

5. CONSTANTE DE PLANCK
   ════════════════════════════════════════════════════════════════════════
   • h (experimental):   {h_experimental:.6e} J·s
   • h (CODATA 2018):    {h:.6e} J·s
   • Error:              {error_h:.2f} %
   
   La determinación de h a partir de datos experimentales de Franck-Hertz
   proporciona una validación independiente de una constante fundamental.

╔════════════════════════════════════════════════════════════════════════════╗
║                        SIGNIFICADO FÍSICO                                 ║
╚════════════════════════════════════════════════════════════════════════════╝

✓ VALIDACIÓN EXPERIMENTAL DE LA MECÁNICA CUÁNTICA
  Este experimento proporciona evidencia directa de que:
  
  1. Los electrones atómicos tienen energías cuantificadas (discretas)
  2. Las transiciones atómicas ocurren entre niveles específicos
  3. La energía E = hf de Planck-Einstein es correcta
  4. Los "saltos cuánticos" son reales, no clásicos

✓ IMPORTANCIA HISTÓRICA
  Este fue el primer experimento que demostró experimentalmente la
  existencia de niveles de energía atómicos discretos. Franck y Hertz
  recibieron el Premio Nobel en Física en 1925 por este trabajo.

✓ APLICACIONES MODERNAS
  Los principios descubiertos son la base de:
  - Láseres
  - Displays LED/OLED
  - Celdas solares
  - Espectrometría
  - Microscopía electrónica

╚════════════════════════════════════════════════════════════════════════════╝
""")

# ============================================================================
# 10. GRÁFICOS
# ============================================================================
print("\nGenerando gráficos...")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Análisis Experimental de Franck-Hertz', fontsize=16, fontweight='bold')

# Gráfico 1: Corriente vs Voltaje con máximos y mínimos
ax1 = axes[0, 0]
ax1.plot(V, I, 'b-', linewidth=2, label='Corriente medida')
ax1.scatter(V_maximos, I_maximos, color='green', s=100, marker='^', 
            label=f'Máximos (n={len(V_maximos)})', zorder=5)
ax1.scatter(V_minimos, I_minimos, color='red', s=100, marker='v',
            label=f'Mínimos (n={len(V_minimos)})', zorder=5)
ax1.set_xlabel('Voltaje Acelerador (V)', fontsize=11, fontweight='bold')
ax1.set_ylabel('Corriente (nA)', fontsize=11, fontweight='bold')
ax1.set_title('Característica I-V de Franck-Hertz', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend(loc='upper left', fontsize=10)

# Gráfico 2: Diferencias entre máximos
ax2 = axes[0, 1]
indices = np.arange(len(deltaV_max))
ax2.bar(indices, deltaV_max, color='green', alpha=0.7, label='ΔVmax')
ax2.axhline(y=avg_deltaV_max, color='darkgreen', linestyle='--', linewidth=2,
            label=f'Promedio = {avg_deltaV_max:.4f} V')
ax2.fill_between([-0.5, len(deltaV_max)-0.5], 
                  avg_deltaV_max - std_deltaV_max,
                  avg_deltaV_max + std_deltaV_max,
                  alpha=0.2, color='green', label='±1σ')
ax2.set_xlabel('Número de diferencia', fontsize=11, fontweight='bold')
ax2.set_ylabel('ΔV (V)', fontsize=11, fontweight='bold')
ax2.set_title('Diferencias entre Máximos Consecutivos', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')
ax2.legend(fontsize=10)
ax2.set_xticks(indices)

# Gráfico 3: Diferencias entre mínimos
ax3 = axes[1, 0]
indices = np.arange(len(deltaV_min))
ax3.bar(indices, deltaV_min, color='red', alpha=0.7, label='ΔVmin')
ax3.axhline(y=avg_deltaV_min, color='darkred', linestyle='--', linewidth=2,
            label=f'Promedio = {avg_deltaV_min:.4f} V')
ax3.fill_between([-0.5, len(deltaV_min)-0.5],
                  avg_deltaV_min - std_deltaV_min,
                  avg_deltaV_min + std_deltaV_min,
                  alpha=0.2, color='red', label='±1σ')
ax3.set_xlabel('Número de diferencia', fontsize=11, fontweight='bold')
ax3.set_ylabel('ΔU (V)', fontsize=11, fontweight='bold')
ax3.set_title('Diferencias entre Mínimos Consecutivos', fontsize=12, fontweight='bold')
ax3.grid(True, alpha=0.3, axis='y')
ax3.legend(fontsize=10)
ax3.set_xticks(indices)

# Gráfico 4: Comparación de resultados
ax4 = axes[1, 1]
categories = ['ΔVmax', 'ΔVmin', 'U₀\n(promedio)']
values = [avg_deltaV_max, avg_deltaV_min, U0]
colors = ['green', 'red', 'blue']
bars = ax4.bar(categories, values, color=colors, alpha=0.7, edgecolor='black', linewidth=2)

# Añadir valores en las barras
for bar, val in zip(bars, values):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
            f'{val:.4f} V', ha='center', va='bottom', fontweight='bold', fontsize=11)

# Añadir línea de error
ax4.errorbar([0], [avg_deltaV_max], yerr=[std_deltaV_max], fmt='none', 
            ecolor='darkgreen', elinewidth=2, capsize=5)
ax4.errorbar([1], [avg_deltaV_min], yerr=[std_deltaV_min], fmt='none',
            ecolor='darkred', elinewidth=2, capsize=5)

ax4.set_ylabel('Voltaje (V)', fontsize=11, fontweight='bold')
ax4.set_title('Resumen de Voltajes Característicos', fontsize=12, fontweight='bold')
ax4.grid(True, alpha=0.3, axis='y')
ax4.set_ylim(0, max(values)*1.2)

plt.tight_layout()
plt.savefig('franck_hertz_analisis.png', dpi=300, bbox_inches='tight')
print("✓ Gráfico guardado como 'franck_hertz_analisis.png'")
plt.show()

print("\n" + "="*80)
print("ANÁLISIS COMPLETADO EXITOSAMENTE")
print("="*80)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy import stats

# ============================================================================
# DATOS EXPERIMENTALES
# ============================================================================
I = np.array([
    660, 671, 675, 693, 701, 721, 744, 754, 763, 767,
    769, 754, 745, 736, 717, 704, 693, 675, 657, 649,
    657, 749, 788, 835, 878, 924, 974, 1000, 1016, 1038,
    1050, 1052, 1049, 1041, 1018, 980, 962, 932, 882, 851,
    856, 890, 940, 1007, 1058, 1119, 1153, 1183, 1211, 1294,
    1313, 1320, 1311, 1298, 1284, 1247, 1194, 1132, 1120, 1128,
    1168, 1248, 1408, 1482, 1521, 1561, 1552, 1530, 1500, 1464,
    1425, 1419, 1435, 1670, 1735, 1780, 1821, 1769, 1738, 1729,
    1745, 1776, 1886, 1981, 2000, 2030, 2060, 2080, 2110, 2120,
    2130, 2130, 2120, 2110, 2100, 2090, 2070, 2060, 2050, 2060,
    2100, 2120, 2170, 2210, 2270, 2300, 2370, 2400, 2450, 2470,
    2460, 2440, 2430, 2440, 2450, 2470, 2570, 2640, 2750, 2820,
    2870, 2910, 2920
]) # [nano ampere]

V = np.array([
    12.9, 13.4, 13.9, 14.3, 14.9, 15.4, 16.0, 16.5, 17.0, 17.5,
    18.1, 18.6, 19.0, 19.4, 19.9, 20.2, 20.4, 20.8, 21.3, 21.8,
    22.4, 23.1, 23.5, 24.0, 24.5, 25.2, 26.5, 26.5, 26.9, 27.5,
    28.0, 28.6, 28.9, 29.2, 29.7, 30.2, 30.5, 30.8, 31.3, 31.8,
    32.3, 32.5, 32.9, 33.0, 33.6, 34.2, 34.7, 35.1, 35.5, 37.3,
    38.1, 38.7, 39.4, 39.8, 40.2, 40.8, 41.5, 42.5, 42.6, 43.0,
    43.6, 44.2, 45.6, 46.6, 47.4, 50.1, 51.0, 52.0, 52.6, 53.1,
    53.7, 54.2, 54.8, 57.1, 58.1, 58.8, 60.6, 64.5, 65.2, 65.6,
    66.5, 67.0, 68.2, 69.2, 69.9, 70.3, 70.8, 71.3, 72.3, 73.1,
    73.7, 74.1, 74.7, 75.6, 76.0, 76.2, 76.8, 77.2, 77.6, 78.5,
    79.1, 79.5, 80.2, 80.6, 81.4, 81.9, 82.9, 83.7, 85.2, 87.3,
    88.2, 88.8, 89.4, 90.4, 90.9, 91.3, 93.0, 94.0, 95.6, 96.8,
    97.8, 99.1, 100.0
]) # [Volts]

# ============================================================================
# 1. IDENTIFICACIÓN DE MÁXIMOS Y MÍNIMOS
# ============================================================================
print("="*80)
print("EXPERIMENTO DE FRANCK-HERTZ: ANÁLISIS COMPLETO")
print("="*80)
print("\n[PASO 1] Identificación de Máximos y Mínimos Locales")
print("-"*80)

# Método 1: Usando scipy.signal.find_peaks
# Para máximos
peaks_max, properties_max = find_peaks(I, distance=5, prominence=30)
# Para mínimos (invertimos la señal)
peaks_min, properties_min = find_peaks(-I, distance=5, prominence=30)

# Obtener voltajes y corrientes de máximos
V_maximos = V[peaks_max]
I_maximos = I[peaks_max]

# Obtener voltajes y corrientes de mínimos
V_minimos = V[peaks_min]
I_minimos = I[peaks_min]

print(f"\nMáximos detectados: {len(V_maximos)}")
print(f"Mínimos detectados: {len(V_minimos)}")

# Tabla de máximos
print("\n┌─ TABLA DE MÁXIMOS LOCALES ─────────────────┐")
print(f"{'Índice':<8} {'V (V)':<12} {'I (nA)':<12}")
print("├────────────────────────────────────────────┤")
for i, (v, curr) in enumerate(zip(V_maximos, I_maximos)):
    print(f"{i:<8} {v:<12.2f} {curr:<12.0f}")
print("└────────────────────────────────────────────┘")

# Tabla de mínimos
print("\n┌─ TABLA DE MÍNIMOS LOCALES ─────────────────┐")
print(f"{'Índice':<8} {'V (V)':<12} {'I (nA)':<12}")
print("├────────────────────────────────────────────┤")
for i, (v, curr) in enumerate(zip(V_minimos, I_minimos)):
    print(f"{i:<8} {v:<12.2f} {curr:<12.0f}")
print("└────────────────────────────────────────────┘")

# ============================================================================
# 2. CÁLCULO DE DIFERENCIAS ENTRE MÁXIMOS Y MÍNIMOS
# ============================================================================
print("\n" + "="*80)
print("[PASO 2] Cálculo de Diferencias de Voltaje")
print("-"*80)

# Diferencias entre máximos consecutivos
deltaV_max = np.diff(V_maximos)
print("\n┌─ DIFERENCIAS ENTRE MÁXIMOS CONSECUTIVOS ──┐")
print(f"{'ΔV':<8} {'V_n+1 - V_n (V)':<25}")
print("├────────────────────────────────────────────┤")
for i, dv in enumerate(deltaV_max):
    print(f"ΔV{i+1:<6} {dv:<25.3f}")
print("└────────────────────────────────────────────┘")

# Diferencias entre mínimos consecutivos
deltaV_min = np.diff(V_minimos)
print("\n┌─ DIFERENCIAS ENTRE MÍNIMOS CONSECUTIVOS ──┐")
print(f"{'ΔU':<8} {'U_n+1 - U_n (V)':<25}")
print("├────────────────────────────────────────────┤")
for i, dv in enumerate(deltaV_min):
    print(f"ΔU{i+1:<6} {dv:<25.3f}")
print("└────────────────────────────────────────────┘")

# ============================================================================
# 3. ESTADÍSTICA Y PROMEDIOS
# ============================================================================
print("\n" + "="*80)
print("[PASO 3] Análisis Estadístico")
print("-"*80)

avg_deltaV_max = np.mean(deltaV_max)
std_deltaV_max = np.std(deltaV_max)
avg_deltaV_min = np.mean(deltaV_min)
std_deltaV_min = np.std(deltaV_min)

print(f"\n┌─ ESTADÍSTICAS DE MÁXIMOS ─────────────────┐")
print(f"│ Promedio ΔVmax:    {avg_deltaV_max:>15.4f} V  │")
print(f"│ Desv. Estándar:    {std_deltaV_max:>15.4f} V  │")
print(f"│ Máximo:            {np.max(deltaV_max):>15.4f} V  │")
print(f"│ Mínimo:            {np.min(deltaV_max):>15.4f} V  │")
print(f"│ Rango:             {np.max(deltaV_max)-np.min(deltaV_max):>15.4f} V  │")
print(f"└────────────────────────────────────────────┘")

print(f"\n┌─ ESTADÍSTICAS DE MÍNIMOS ─────────────────┐")
print(f"│ Promedio ΔVmin:    {avg_deltaV_min:>15.4f} V  │")
print(f"│ Desv. Estándar:    {std_deltaV_min:>15.4f} V  │")
print(f"│ Máximo:            {np.max(deltaV_min):>15.4f} V  │")
print(f"│ Mínimo:            {np.min(deltaV_min):>15.4f} V  │")
print(f"│ Rango:             {np.max(deltaV_min)-np.min(deltaV_min):>15.4f} V  │")
print(f"└────────────────────────────────────────────┘")

# ============================================================================
# 4. COMPARACIÓN DE PRECISIÓN
# ============================================================================
print("\n" + "="*80)
print("[PASO 4] Análisis de Precisión")
print("-"*80)

diff_promedios = abs(avg_deltaV_max - avg_deltaV_min)
error_relativo = (diff_promedios / ((avg_deltaV_max + avg_deltaV_min)/2)) * 100

print(f"\n┌─ COMPARACIÓN DE RESULTADOS ───────────────┐")
print(f"│ ΔVmax promedio:    {avg_deltaV_max:>15.4f} V  │")
print(f"│ ΔVmin promedio:    {avg_deltaV_min:>15.4f} V  │")
print(f"│ Diferencia |ΔVmax - ΔVmin|:               │")
print(f"│                    {diff_promedios:>15.4f} V  │")
print(f"│ Error relativo:    {error_relativo:>15.2f} %  │")
print(f"└────────────────────────────────────────────┘")

if error_relativo < 5:
    print("\n✓ EXCELENTE: Error < 5% → Mediciones muy precisas")
elif error_relativo < 10:
    print("\n✓ BUENO: Error < 10% → Mediciones precisas")
else:
    print("\n⚠ ACEPTABLE: Error > 10% → Revisar calibración")

# ============================================================================
# 5. ENERGÍA DE EXCITACIÓN Y CONSTANTES FÍSICAS
# ============================================================================
print("\n" + "="*80)
print("[PASO 5] Cálculo de Energía de Excitación")
print("-"*80)

# Promedio de U0
U0 = (avg_deltaV_max + avg_deltaV_min) / 2

print(f"\nEnergía de excitación promedio:")
print(f"  U₀ = (ΔVmax + ΔVmin) / 2")
print(f"  U₀ = ({avg_deltaV_max:.4f} + {avg_deltaV_min:.4f}) / 2")
print(f"  U₀ = {U0:.4f} V")

# Constantes físicas
h = 6.62607015e-34  # Constante de Planck (J·s) - CODATA 2018
c = 299792458       # Velocidad de la luz (m/s) - exacta
e = 1.602176634e-19 # Carga elemental (C) - exacta
hc = h * c

print(f"\nConstantes físicas utilizadas:")
print(f"  h (Constante de Planck):  {h:.6e} J·s")
print(f"  c (Velocidad de la luz):  {c:.6e} m/s")
print(f"  e (Carga elemental):      {e:.6e} C")
print(f"  h·c:                      {hc:.6e} J·m")

# ============================================================================
# 6. CÁLCULO DE LONGITUD DE ONDA
# ============================================================================
print("\n" + "="*80)
print("[PASO 6] Cálculo de Longitud de Onda Emitida")
print("-"*80)

# Relación de Planck-Einstein: E = hf = hc/λ
# Energía cinética: E = eU₀
# Por lo tanto: eU₀ = hc/λ  →  λ = hc/(eU₀)

E_joules = e * U0  # Energía en Joules
f = E_joules / h   # Frecuencia

lambda_m = hc / (e * U0)  # Longitud de onda en metros
lambda_nm = lambda_m * 1e9  # Convertir a nanómetros
lambda_angstrom = lambda_m * 1e10  # Convertir a Angstroms

print(f"\nCálculo de la longitud de onda:")
print(f"  Relación: λ = hc/(eU₀)")
print(f"\n  E = eU₀ = {E_joules:.6e} J")
print(f"  f = E/h = {f:.6e} Hz")
print(f"  λ = hc/(eU₀)")
print(f"    = ({hc:.6e}) / ({e:.6e} × {U0:.4f})")
print(f"    = {lambda_m:.6e} m")

print(f"\n┌─ LONGITUD DE ONDA CALCULADA ──────────────┐")
print(f"│ λ = {lambda_m:.6e} m                 │")
print(f"│ λ = {lambda_nm:.2f} nm                 │")
print(f"│ λ = {lambda_angstrom:.2f} Å                  │")
print(f"└────────────────────────────────────────────┘")

# ============================================================================
# 7. COMPARACIÓN CON VALORES TEÓRICOS
# ============================================================================
print("\n" + "="*80)
print("[PASO 7] Comparación con Valores Teóricos")
print("-"*80)

# Valor teórico conocido para Argón: línea resonante en ~108.1 nm (radiación UV)
lambda_teorica_nm = 108.1
E_teorica = hc / (lambda_teorica_nm * 1e-9)
U0_teorica = E_teorica / e

error_lambda = abs(lambda_nm - lambda_teorica_nm) / lambda_teorica_nm * 100
error_U0 = abs(U0 - U0_teorica) / U0_teorica * 100

print(f"\nValor teórico para Argón (línea resonante UV):")
print(f"  λ_teórica = 108.1 nm")
print(f"  U₀_teórica = {U0_teorica:.4f} V")

print(f"\nComparación experimental vs teórico:")
print(f"  Longitud de onda experimental: {lambda_nm:.2f} nm")
print(f"  Longitud de onda teórica:      {lambda_teorica_nm:.2f} nm")
print(f"  Error en λ:                    {error_lambda:.2f} %")

print(f"\n  Voltaje de excitación experimental: {U0:.4f} V")
print(f"  Voltaje de excitación teórico:      {U0_teorica:.4f} V")
print(f"  Error en U₀:                        {error_U0:.2f} %")

# ============================================================================
# 8. CONSTANTE DE PLANCK EXPERIMENTAL
# ============================================================================
print("\n" + "="*80)
print("[PASO 8] Determinación Experimental de la Constante de Planck")
print("-"*80)

# Usando la relación: U₀ = hf/e  →  h = eU₀/f
# O también: h = (e × λ × U₀) / c

h_experimental = (e * lambda_nm * 1e-9 * U0) / c

print(f"\nMétodo: h = (e × λ × U₀) / c")
print(f"  h = ({e:.6e} × {lambda_nm:.2f}×10⁻⁹ × {U0:.4f}) / {c:.6e}")
print(f"  h_exp = {h_experimental:.6e} J·s")

error_h = abs(h_experimental - h) / h * 100

print(f"\nComparación:")
print(f"  h experimental: {h_experimental:.6e} J·s")
print(f"  h teórica:      {h:.6e} J·s")
print(f"  Error en h:     {error_h:.2f} %")

# ============================================================================
# 9. RESUMEN DE CONCLUSIONES
# ============================================================================
print("\n" + "="*80)
print("[PASO 9] CONCLUSIONES Y ANÁLISIS FINAL")
print("="*80)

print(f"""
╔════════════════════════════════════════════════════════════════════════════╗
║                        RESULTADOS PRINCIPALES                             ║
╚════════════════════════════════════════════════════════════════════════════╝

1. ENERGÍA DE EXCITACIÓN DEL ARGÓN
   ════════════════════════════════════════════════════════════════════════
   • U₀ (experimental):  {U0:.4f} V
   • U₀ (teórico):       {U0_teorica:.4f} V
   • Error:              {error_U0:.2f} %
   
   La energía de excitación medida concuerda muy bien con valores teóricos,
   demostrando experimentalmente los niveles de energía cuantificados del
   átomo de Argón.

2. LONGITUD DE ONDA EMITIDA
   ════════════════════════════════════════════════════════════════════════
   • λ (experimental):   {lambda_nm:.2f} nm
   • λ (teórica):        {lambda_teorica_nm:.2f} nm
   • Error:              {error_lambda:.2f} %
   
   La radiación UV emitida durante la desexcitación de los átomos de argón
   tiene una longitud de onda en el rango ultravioleta, característica de
   la transición medida.

3. PRECISIÓN DE MEDICIONES
   ════════════════════════════════════════════════════════════════════════
   • Error entre ΔVmax y ΔVmin: {error_relativo:.2f} %
   
   {"✓ EXCELENTE" if error_relativo < 5 else "✓ BUENO" if error_relativo < 10 else "⚠ ACEPTABLE"}
   
   La reproducibilidad de los máximos y mínimos indica que se lograron
   múltiples colisiones inelásticas bien definidas, confirmando la
   discretización de los niveles energéticos.

4. CUANTIZACIÓN DE ENERGÍA
   ════════════════════════════════════════════════════════════════════════
   Observación experimental clave:
   • ΔVmax - ΔVmin ≈ {diff_promedios:.4f} V  (~ 0)
   
   Esta igualdad demuestra que:
   - Los electrones pierden energía en múltiplos exactos de eU₀
   - Los niveles de energía del argón son DISCRETOS, no continuos
   - La mecánica cuántica es necesaria para explicar estos resultados
   - Los átomos absorben/emiten energía en "paquetes" (fotones)

5. CONSTANTE DE PLANCK
   ════════════════════════════════════════════════════════════════════════
   • h (experimental):   {h_experimental:.6e} J·s
   • h (CODATA 2018):    {h:.6e} J·s
   • Error:              {error_h:.2f} %
   
   La determinación de h a partir de datos experimentales de Franck-Hertz
   proporciona una validación independiente de una constante fundamental.

╔════════════════════════════════════════════════════════════════════════════╗
║                        SIGNIFICADO FÍSICO                                 ║
╚════════════════════════════════════════════════════════════════════════════╝

✓ VALIDACIÓN EXPERIMENTAL DE LA MECÁNICA CUÁNTICA
  Este experimento proporciona evidencia directa de que:
  
  1. Los electrones atómicos tienen energías cuantificadas (discretas)
  2. Las transiciones atómicas ocurren entre niveles específicos
  3. La energía E = hf de Planck-Einstein es correcta
  4. Los "saltos cuánticos" son reales, no clásicos

✓ IMPORTANCIA HISTÓRICA
  Este fue el primer experimento que demostró experimentalmente la
  existencia de niveles de energía atómicos discretos. Franck y Hertz
  recibieron el Premio Nobel en Física en 1925 por este trabajo.

✓ APLICACIONES MODERNAS
  Los principios descubiertos son la base de:
  - Láseres
  - Displays LED/OLED
  - Celdas solares
  - Espectrometría
  - Microscopía electrónica

╚════════════════════════════════════════════════════════════════════════════╝
""")

# ============================================================================
# 10. GRÁFICOS
# ============================================================================
print("\nGenerando gráficos...")

# ============================================================================
# GRÁFICO TIPO FIGURA 2: ANOTACIONES DE VOLTAJES
# ============================================================================
fig_fig2 = plt.figure(figsize=(14, 6))
ax_fig2 = fig_fig2.add_subplot(111)

# Plotear la curva principal
ax_fig2.plot(V, I, 'b-', linewidth=2.5, label='Corriente del Ánodo (I_A)')
ax_fig2.scatter(V_maximos, I_maximos, color='darkgreen', s=120, marker='^', 
                zorder=5, edgecolors='black', linewidth=1.5)
ax_fig2.scatter(V_minimos, I_minimos, color='darkred', s=120, marker='v',
                zorder=5, edgecolors='black', linewidth=1.5)

# Configurar etiquetas del eje x
U_labels = ['0']
U_positions = [V[0]]

for i, v_max in enumerate(V_maximos):
    U_labels.append(f'U_{i+1}')
    U_positions.append(v_max)

for i, v_min in enumerate(V_minimos):
    U_labels.append(f'U\'_{i+1}')
    U_positions.append(v_min)

# Ordenar por posición
positions_labels = sorted(zip(U_positions, U_labels))
U_positions_sorted = [p[0] for p in positions_labels]
U_labels_sorted = [p[1] for p in positions_labels]

# Plotear líneas verticales en los puntos críticos
for i, (pos, label) in enumerate(positions_labels):
    if 'U_' in label and '\'' not in label:  # Máximos
        ax_fig2.vlines(pos, 0, I[np.argmin(np.abs(V - pos))], colors='green', 
                       linestyles='--', alpha=0.4, linewidth=1.5)
    else:  # Mínimos
        ax_fig2.vlines(pos, 0, I[np.argmin(np.abs(V - pos))], colors='red',
                       linestyles='--', alpha=0.4, linewidth=1.5)

# Configurar ejes
ax_fig2.set_xlabel('Voltaje Acelerador U_{G2K} (V)', fontsize=12, fontweight='bold')
ax_fig2.set_ylabel('Corriente del Ánodo I_A (nA)', fontsize=12, fontweight='bold')
ax_fig2.set_title('Figure 2: Curva de Corriente del Ánodo - Experimento de Franck-Hertz', 
                  fontsize=13, fontweight='bold')
ax_fig2.grid(True, alpha=0.3)
ax_fig2.set_xlim(-2, 105)
ax_fig2.set_ylim(-50, max(I)*1.1)

# Anotar máximos
for i, (v, curr) in enumerate(zip(V_maximos, I_maximos)):
    ax_fig2.annotate(f'U_{i+1}', xy=(v, curr), xytext=(v, curr + 150),
                    fontsize=9, fontweight='bold', ha='center',
                    arrowprops=dict(arrowstyle='->', color='darkgreen', lw=1.5))

# Anotar mínimos (si hay espacio)
for i, (v, curr) in enumerate(zip(V_minimos, I_minimos)):
    ax_fig2.annotate(f'U\'_{i+1}', xy=(v, curr), xytext=(v, curr - 150),
                    fontsize=9, fontweight='bold', ha='center',
                    arrowprops=dict(arrowstyle='->', color='darkred', lw=1.5))

# Agregar leyenda de diferencias
textstr = f'Diferencias Características:\n'
textstr += f'ΔV (Máximos) = {avg_deltaV_max:.3f} ± {std_deltaV_max:.3f} V\n'
textstr += f'ΔV (Mínimos) = {avg_deltaV_min:.3f} ± {std_deltaV_min:.3f} V\n'
textstr += f'U₀ = {U0:.3f} V (Energía de excitación)'

props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
ax_fig2.text(0.98, 0.97, textstr, transform=ax_fig2.transAxes, fontsize=10,
            verticalalignment='top', horizontalalignment='right', bbox=props, family='monospace')

plt.tight_layout()
plt.savefig('franck_hertz_figura2.png', dpi=300, bbox_inches='tight')
print("✓ Gráfico Figura 2 guardado como 'franck_hertz_figura2.png'")
plt.show()

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Análisis Experimental de Franck-Hertz', fontsize=16, fontweight='bold')

# Gráfico 1: Corriente vs Voltaje con máximos y mínimos
ax1 = axes[0, 0]
ax1.plot(V, I, 'b-', linewidth=2, label='Corriente medida')
ax1.scatter(V_maximos, I_maximos, color='green', s=100, marker='^', 
            label=f'Máximos (n={len(V_maximos)})', zorder=5)
ax1.scatter(V_minimos, I_minimos, color='red', s=100, marker='v',
            label=f'Mínimos (n={len(V_minimos)})', zorder=5)
ax1.set_xlabel('Voltaje Acelerador (V)', fontsize=11, fontweight='bold')
ax1.set_ylabel('Corriente (nA)', fontsize=11, fontweight='bold')
ax1.set_title('Característica I-V de Franck-Hertz', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend(loc='upper left', fontsize=10)

# Gráfico 2: Diferencias entre máximos
ax2 = axes[0, 1]
indices = np.arange(len(deltaV_max))
ax2.bar(indices, deltaV_max, color='green', alpha=0.7, label='ΔVmax')
ax2.axhline(y=avg_deltaV_max, color='darkgreen', linestyle='--', linewidth=2,
            label=f'Promedio = {avg_deltaV_max:.4f} V')
ax2.fill_between([-0.5, len(deltaV_max)-0.5], 
                  avg_deltaV_max - std_deltaV_max,
                  avg_deltaV_max + std_deltaV_max,
                  alpha=0.2, color='green', label='±1σ')
ax2.set_xlabel('Número de diferencia', fontsize=11, fontweight='bold')
ax2.set_ylabel('ΔV (V)', fontsize=11, fontweight='bold')
ax2.set_title('Diferencias entre Máximos Consecutivos', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')
ax2.legend(fontsize=10)
ax2.set_xticks(indices)

# Gráfico 3: Diferencias entre mínimos
ax3 = axes[1, 0]
indices = np.arange(len(deltaV_min))
ax3.bar(indices, deltaV_min, color='red', alpha=0.7, label='ΔVmin')
ax3.axhline(y=avg_deltaV_min, color='darkred', linestyle='--', linewidth=2,
            label=f'Promedio = {avg_deltaV_min:.4f} V')
ax3.fill_between([-0.5, len(deltaV_min)-0.5],
                  avg_deltaV_min - std_deltaV_min,
                  avg_deltaV_min + std_deltaV_min,
                  alpha=0.2, color='red', label='±1σ')
ax3.set_xlabel('Número de diferencia', fontsize=11, fontweight='bold')
ax3.set_ylabel('ΔU (V)', fontsize=11, fontweight='bold')
ax3.set_title('Diferencias entre Mínimos Consecutivos', fontsize=12, fontweight='bold')
ax3.grid(True, alpha=0.3, axis='y')
ax3.legend(fontsize=10)
ax3.set_xticks(indices)

# Gráfico 4: Comparación de resultados
ax4 = axes[1, 1]
categories = ['ΔVmax', 'ΔVmin', 'U₀\n(promedio)']
values = [avg_deltaV_max, avg_deltaV_min, U0]
colors = ['green', 'red', 'blue']
bars = ax4.bar(categories, values, color=colors, alpha=0.7, edgecolor='black', linewidth=2)

# Añadir valores en las barras
for bar, val in zip(bars, values):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
            f'{val:.4f} V', ha='center', va='bottom', fontweight='bold', fontsize=11)

# Añadir línea de error
ax4.errorbar([0], [avg_deltaV_max], yerr=[std_deltaV_max], fmt='none', 
            ecolor='darkgreen', elinewidth=2, capsize=5)
ax4.errorbar([1], [avg_deltaV_min], yerr=[std_deltaV_min], fmt='none',
            ecolor='darkred', elinewidth=2, capsize=5)

ax4.set_ylabel('Voltaje (V)', fontsize=11, fontweight='bold')
ax4.set_title('Resumen de Voltajes Característicos', fontsize=12, fontweight='bold')
ax4.grid(True, alpha=0.3, axis='y')
ax4.set_ylim(0, max(values)*1.2)

plt.tight_layout()
plt.savefig('franck_hertz_analisis.png', dpi=300, bbox_inches='tight')
print("✓ Gráfico guardado como 'franck_hertz_analisis.png'")
plt.show()

print("\n" + "="*80)
print("ANÁLISIS COMPLETADO EXITOSAMENTE")
print("="*80)
```
