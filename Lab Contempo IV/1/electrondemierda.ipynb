```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Datos proporcionados
U = np.array([300, 280, 260, 240, 220, 200])  # [V]
R = np.array([0.03, 0.04, 0.05])  # [m]
I_h3 = np.array([2.53, 2.34, 2.21, 2.09, 1.89, 1.74])  # [A] para r = 3 cm
I_h4 = np.array([1.91, 1.82, 1.67, 1.57, 1.44, 1.34])  # [A] para r = 4 cm
I_h5 = np.array([1.54, 1.49, 1.41, 1.33, 1.23, 1.12])  # [A] para r = 5 cm
k = 0.756 * 1e-3  # [T/A] (convertido de mT a T)

# Listas para almacenar todos los datos combinados
all_rB_squared = []
all_2U = []

# Función para procesar cada conjunto de datos
def procesar_datos(r, I_h, radio_nombre):
    print(f"\n=== Radio {radio_nombre} ===")
    B = k * I_h  # Campo magnético [T]
    rB = r * B  # r * B [m·T]
    rB_squared = (rB)**2  # (r·B)^2 [m²·T²]
    dosU = 2 * U  # 2U [V]
    
    print(f"U (V): {U}")
    print(f"I_h (A): {I_h}")
    print(f"B (T): {B}")
    print(f"r·B (m·T): {rB}")
    print(f"(r·B)^2 (m²·T²): {rB_squared}")
    print(f"2U (V): {dosU}")
    
    return rB_squared, dosU

# Procesar datos para cada radio
rB_sq_3cm, dosU_3cm = procesar_datos(R[0], I_h3, "3 cm")
rB_sq_4cm, dosU_4cm = procesar_datos(R[1], I_h4, "4 cm")
rB_sq_5cm, dosU_5cm = procesar_datos(R[2], I_h5, "5 cm")

# Combinar todos los datos
all_rB_squared = np.concatenate([rB_sq_3cm, rB_sq_4cm, rB_sq_5cm])
all_2U = np.concatenate([dosU_3cm, dosU_4cm, dosU_5cm])

# Realizar regresión lineal
slope, intercept, r_value, p_value, std_err = linregress(all_rB_squared, all_2U)

# La pendiente es e/m
e_m = slope

# Valor teórico de referencia
e_m_teorico = 1.7588e11  # C/kg

# Calcular error porcentual
error_porcentual = abs((e_m - e_m_teorico) / e_m_teorico) * 100

# Crear gráfico
plt.figure(figsize=(10, 6))
plt.scatter(all_rB_squared, all_2U, alpha=0.7, label='Datos experimentales')

# Línea de regresión
x_fit = np.linspace(min(all_rB_squared), max(all_rB_squared), 100)
y_fit = slope * x_fit + intercept
plt.plot(x_fit, y_fit, 'r-', label=f'Regresión lineal: y = {slope:.2e}x + {intercept:.2f}')

plt.xlabel('$(r \\cdot B)^2$ [m²·T²]')
plt.ylabel('2U [V]')
plt.title('Determinación de $e/m$ del electrón\n$2U$ vs $(r \\cdot B)^2$')
plt.grid(True, alpha=0.3)
plt.legend()

# Añadir texto con los resultados
textstr = '\n'.join((
    f'$e/m$ experimental = {e_m:.3e} C/kg',
    f'$e/m$ teórico = {e_m_teorico:.3e} C/kg',
    f'Error = {error_porcentual:.2f}%',
    f'$R^2$ = {r_value**2:.4f}'))
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
plt.text(0.05, 0.95, textstr, transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()

# Resultados detallados
print(f"\n=== RESULTADOS FINALES ===")
print(f"Pendiente (e/m) = {e_m:.3e} C/kg")
print(f"Intercepto = {intercept:.3f} V")
print(f"Coeficiente de correlación (R) = {r_value:.6f}")
print(f"R² = {r_value**2:.6f}")
print(f"Error estándar = {std_err:.3e}")
print(f"Valor teórico e/m = {e_m_teorico:.3e} C/kg")
print(f"Error porcentual = {error_porcentual:.2f}%")

# Resultados por separado para cada radio
print(f"\n=== RESULTADOS POR RADIO ===")
for i, (r_val, I_h, radio_nombre) in enumerate(zip(
    [R[0], R[1], R[2]], 
    [I_h3, I_h4, I_h5], 
    ['3 cm', '4 cm', '5 cm']
)):
    B = k * I_h
    rB_squared = (r_val * B)**2
    dosU = 2 * U
    
    slope_individual, intercept_individual, r_value_individual, _, _ = linregress(rB_squared, dosU)
    error_individual = abs((slope_individual - e_m_teorico) / e_m_teorico) * 100
    
    print(f"Radio {radio_nombre}:")
    print(f"  e/m = {slope_individual:.3e} C/kg")
    print(f"  Error = {error_individual:.2f}%")
    print(f"  R² = {r_value_individual**2:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Datos proporcionados
U = np.array([300, 280, 260, 240, 220, 200])  # [V]
R = np.array([0.03, 0.04, 0.05])  # [m]
I_h3 = np.array([2.53, 2.34, 2.21, 2.09, 1.89, 1.74])  # [A] para r = 3 cm
I_h4 = np.array([1.91, 1.82, 1.67, 1.57, 1.44, 1.34])  # [A] para r = 4 cm
I_h5 = np.array([1.54, 1.49, 1.41, 1.33, 1.23, 1.12])  # [A] para r = 5 cm
k = 0.756  # [mT/A] = [10^-3 T/A]

# Convertir k a Tesla/Ampere
k_T = k * 1e-3  # [T/A]

# Calcular B para cada corriente
B3 = k_T * I_h3  # [T] para r = 3 cm
B4 = k_T * I_h4  # [T] para r = 4 cm
B5 = k_T * I_h5  # [T] para r = 5 cm

# Calcular r²B² para cada radio
r3_sq_B3_sq = (R[0]**2) * (B3**2)  # [m²·T²] para r = 3 cm
r4_sq_B4_sq = (R[1]**2) * (B4**2)  # [m²·T²] para r = 4 cm
r5_sq_B5_sq = (R[2]**2) * (B5**2)  # [m²·T²] para r = 5 cm

# Calcular 2U
U_double = 2 * U  # [V]

# Combinar todos los datos para el ajuste lineal
all_r2B2 = np.concatenate([r3_sq_B3_sq, r4_sq_B4_sq, r5_sq_B5_sq])
all_2U = np.concatenate([U_double, U_double, U_double])

# Ajuste lineal: 2U vs r²B²
slope, intercept, r_value, p_value, std_err = linregress(all_r2B2, all_2U)

# La pendiente es e/m según la ecuación (5)
e_m = slope  # [C/kg]

# Valor teórico de e/m
e_m_theoretical = 1.758820e11  # [C/kg]

# Calcular el error porcentual
error_percent = abs((e_m - e_m_theoretical) / e_m_theoretical) * 100

# Resultados
print("RESULTADOS DEL EXPERIMENTO")
print("=" * 50)
print(f"Pendiente del ajuste lineal: {slope:.3e} C/kg")
print(f"Relación carga/masa experimental (e/m): {e_m:.3e} C/kg")
print(f"Relación carga/masa teórica (e/m): {e_m_theoretical:.3e} C/kg")
print(f"Error porcentual: {error_percent:.2f}%")
print(f"Coeficiente de correlación (R²): {r_value**2:.4f}")

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(all_r2B2, all_2U, alpha=0.7, label='Datos experimentales')
plt.plot(all_r2B2, slope * all_r2B2 + intercept, 'r-')

plt.xlabel('r²B² [m²·T²]')
plt.ylabel('2U [V]')
plt.title('Determinación de e/m: 2U vs r²B²')
plt.grid(True, alpha=0.3)
plt.legend()

# Añadir texto con el resultado
textstr = f'e/m = {e_m:.3e} C/kg\nR² = {r_value**2:.4f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
plt.text(0.05, 0.95, textstr, transform=plt.gca().transAxes, fontsize=12,
         verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()

# Resultados detallados por radio
print("\nRESULTADOS POR RADIO:")
print("=" * 50)
print("Para r = 3 cm:")
for i, u in enumerate(U):
    print(f"U = {u} V, I = {I_h3[i]} A, B = {B3[i]:.4f} T, r²B² = {r3_sq_B3_sq[i]:.6e} m²T²")

print("\nPara r = 4 cm:")
for i, u in enumerate(U):
    print(f"U = {u} V, I = {I_h4[i]} A, B = {B4[i]:.4f} T, r²B² = {r4_sq_B4_sq[i]:.6e} m²T²")

print("\nPara r = 5 cm:")
for i, u in enumerate(U):
    print(f"U = {u} V, I = {I_h5[i]} A, B = {B5[i]:.4f} T, r²B² = {r5_sq_B5_sq[i]:.6e} m²T²")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Datos proporcionados
U = np.array([300, 280, 260, 240, 220, 200])  # [V]
R = np.array([0.026, 0.04, 0.05])  # [m]
I_h3 = np.array([2.53, 2.34, 2.21, 2.09, 1.89, 1.7])  # [A] para r = 3 cm
I_h4 = np.array([1.91, 1.82, 1.67, 1.57, 1.44, 1.34])  # [A] para r = 4 cm
I_h5 = np.array([1.54, 1.49, 1.41, 1.33, 1.23, 1.12])  # [A] para r = 5 cm
k = 0.756  # [mT/A] = [10^-3 T/A]

# Convertir k a Tesla/Ampere
k_T = k * 1e-3  # [T/A]

# Calcular B para cada corriente
B3 = k_T * I_h3  # [T] para r = 3 cm
B4 = k_T * I_h4  # [T] para r = 4 cm
B5 = k_T * I_h5  # [T] para r = 5 cm

# Calcular r²B² para cada radio
r3_sq_B3_sq = (R[0]**2) * (B3**2)  # [m²·T²] para r = 3 cm
r4_sq_B4_sq = (R[1]**2) * (B4**2)  # [m²·T²] para r = 4 cm
r5_sq_B5_sq = (R[2]**2) * (B5**2)  # [m²·T²] para r = 5 cm

# Calcular 2U
U_double = 2 * U  # [V]

# Combinar todos los datos para el ajuste lineal
all_r2B2 = np.concatenate([r3_sq_B3_sq, r4_sq_B4_sq, r5_sq_B5_sq])
all_2U = np.concatenate([U_double, U_double, U_double])

# Ajuste lineal: 2U vs r²B²
slope, intercept, r_value, p_value, std_err = linregress(all_r2B2, all_2U)

# La pendiente es e/m según la ecuación (5)
e_m_exp = slope  # [C/kg]
e_m_error = std_err  # Error estándar de la pendiente

# Valor teórico de e/m
e_m_theoretical = 1.758820e11  # [C/kg]

# Calcular errores
error_abs = abs(e_m_exp - e_m_theoretical)
error_percent = (error_abs / e_m_theoretical) * 100
error_relative = error_abs / e_m_theoretical

# Calcular incertidumbre porcentual
uncertainty_percent = (e_m_error / e_m_exp) * 100

# Resultados
print("RESULTADOS DEL EXPERIMENTO")
print("=" * 60)
print(f"Pendiente del ajuste lineal: ({slope:.3e} ± {std_err:.3e}) C/kg")
print(f"Relación carga/masa experimental (e/m): ({e_m_exp:.3e} ± {e_m_error:.3e}) C/kg") 
print(f"Relación carga/masa teórica (e/m): {e_m_theoretical:.3e} C/kg")
print(f"Error absoluto: {error_abs:.3e} C/kg")
print(f"Error porcentual: {error_percent:.2f}%")
print(f"Error relativo: {error_relative:.4f}")
print(f"Incertidumbre experimental: ±{uncertainty_percent:.2f}%")
print(f"Coeficiente de correlación (R²): {r_value**2:.6f}")
print(f"Intercepto: {intercept:.3e} V (teóricamente debería ser 0)")

# Gráfico
plt.figure(figsize=(12, 8))

# Datos experimentales por radio
plt.scatter(r3_sq_B3_sq, U_double, color='blue', alpha=0.7, label='r = 3 cm', s=80)
plt.scatter(r4_sq_B4_sq, U_double, color='red', alpha=0.7, label='r = 4 cm', s=80)
plt.scatter(r5_sq_B5_sq, U_double, color='green', alpha=0.7, label='r = 5 cm', s=80)

# Ajuste lineal
x_fit = np.linspace(min(all_r2B2), max(all_r2B2), 100)
y_fit = slope * x_fit + intercept
plt.plot(x_fit, y_fit, 'k-', linewidth=2, label=f'Ajuste lineal: 2U = {slope:.3e}·r²B² + {intercept:.3e}')

# Línea teórica (si el intercepto fuera 0 perfecto)
y_theoretical = e_m_theoretical * x_fit
plt.plot(x_fit, y_theoretical, 'r--', linewidth=2, label=f'Valor teórico: 2U = {e_m_theoretical:.3e}·r²B²')

plt.xlabel('r²B² [m²·T²]', fontsize=12)
plt.ylabel('2U [V]', fontsize=12)
plt.title('Determinación de la relación carga/masa del electrón (e/m)', fontsize=14)
plt.grid(True, alpha=0.3)

# Texto con resultados
textstr = '\n'.join([
    f'e/m experimental = ({e_m_exp:.3e} ± {e_m_error:.3e}) C/kg',
    f'e/m teórico = {e_m_theoretical:.3e} C/kg',
    f'Error = {error_percent:.2f}%',
    f'R² = {r_value**2:.6f}',
    f'Incertidumbre = ±{uncertainty_percent:.2f}%'
])

props = dict(boxstyle='round', facecolor='lightblue', alpha=0.9)
plt.text(0.02, 0.98, textstr, transform=plt.gca().transAxes, fontsize=11,
         verticalalignment='top', bbox=props)

plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

# Análisis estadístico adicional
print("\nANÁLISIS ESTADÍSTICO DETALLADO")
print("=" * 60)

# Calcular residuos
residuals = all_2U - (slope * all_r2B2 + intercept)
rmse = np.sqrt(np.mean(residuals**2))
mean_2U = np.mean(all_2U)
rms_percent_error = (rmse / mean_2U) * 100

print(f"Número de puntos: {len(all_2U)}")
print(f"RMSE (Root Mean Square Error): {rmse:.3f} V")
print(f"Error RMS porcentual: {rms_percent_error:.2f}%")
print(f"Rango de 2U: {min(all_2U)} - {max(all_2U)} V")
print(f"Rango de r²B²: {min(all_r2B2):.3e} - {max(all_r2B2):.3e} m²T²")

# Resultados por radio individual
print("\nRESULTADOS POR RADIO INDIVIDUAL:")
print("=" * 60)

for i, (radius, currents, B_values, r2B2_values) in enumerate(zip(
    [0.03, 0.04, 0.05], 
    [I_h3, I_h4, I_h5], 
    [B3, B4, B5], 
    [r3_sq_B3_sq, r4_sq_B4_sq, r5_sq_B5_sq]
), 1):
    print(f"\nRadio {i}: r = {radius*100} cm")
    print("-" * 30)
    for j, u in enumerate(U):
        e_m_individual = (2 * u) / (r2B2_values[j])
        error_ind = abs(e_m_individual - e_m_theoretical) / e_m_theoretical * 100
        print(f"U = {u:3d} V: I = {currents[j]:.2f} A, B = {B_values[j]:.4f} T, "
              f"e/m = {e_m_individual:.3e} C/kg, Error = {error_ind:.1f}%")
    
    # Ajuste lineal individual para cada radio
    slope_ind, intercept_ind, r_value_ind, _, _ = linregress(r2B2_values, 2*U)
    error_ind_avg = abs(slope_ind - e_m_theoretical) / e_m_theoretical * 100
    print(f"Ajuste individual: e/m = {slope_ind:.3e} C/kg, Error = {error_ind_avg:.1f}%, R² = {r_value_ind**2:.4f}")
```
