```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Datos experimentales
voltaje = np.array([-0.55, -0.74, -1.29, -1.53, -2.13])
longitud_onda_m = np.array([5.77e-7, 5.46e-7, 4.36e-7, 4.05e-7, 3.65e-7])

# Constantes fsicas
c = 3e8  # velocidad de la luz en m/s
e = 1.602e-19  # carga del electrn en C
h_teorico = 6.626e-34  # valor terico de Planck en Js

# Clculos
frecuencia = c / longitud_onda_m  # en Hz
energia = e * np.abs(voltaje)  # en Joules

# Ajuste lineal
pendiente, intercepto, r_value, p_value, std_err = stats.linregress(frecuencia, energia)
h_experimental = pendiente

# Predicciones del modelo
frecuencia_ajuste = np.linspace(min(frecuencia), max(frecuencia), 100)
energia_ajuste = pendiente * frecuencia_ajuste + intercepto

# Clculo de errores
error_absoluto = abs(h_experimental - h_teorico)
error_porcentual = (error_absoluto / h_teorico) * 100

# Trabajo de extraccin
trabajo_extraccion = -intercepto  # en Joules
trabajo_extraccion_ev = trabajo_extraccion / e  # en eV

# Grfico
plt.figure(figsize=(10, 6))
plt.plot(frecuencia, energia, 'bo', markersize=8, label='Datos experimentales')
plt.plot(frecuencia_ajuste, energia_ajuste, 'r-', linewidth=2,
         label=f'Ajuste lineal: E = hf - \nh = {h_experimental:.3e} Js')

plt.xlabel('Frecuencia (Hz)', fontsize=12)
plt.ylabel('Energa (J)', fontsize=12)
plt.title('Determinacin de la Constante de Planck - Efecto Fotoelctrico', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)

# Anotaciones en el grfico
plt.annotate(f'h experimental = {h_experimental:.3e} Js\n'
             f'h terico = {h_teorico:.3e} Js\n'
             f'Error = {error_porcentual:.2f}%',
             xy=(0.05, 0.75), xycoords='axes fraction',
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8),
             fontsize=10)

plt.tight_layout()
plt.show()

# Resultados detallados
print("=" * 60)
print("ANLISIS DEL EFECTO FOTOELCTRICO")
print("=" * 60)

print("\nDATOS PROCESADOS:")
print("-" * 40)
print("V (V)     (m)        f (Hz)        E (J)")
print("-" * 40)
for i in range(len(voltaje)):
    print(f"{voltaje[i]:.2f}   {longitud_onda_m[i]:.2e}   {frecuencia[i]:.2e}   {energia[i]:.2e}")

print(f"\nRESULTADOS DEL AJUSTE LINEAL:")
print("-" * 40)
print(f"Constante de Planck experimental: {h_experimental:.4e} Js")
print(f"Constante de Planck terica:      {h_teorico:.4e} Js")
print(f"Error absoluto:                   {error_absoluto:.4e} Js")
print(f"Error porcentual:                 {error_porcentual:.2f} %")
print(f"Coeficiente de correlacin (r):   {r_value:.4f}")
print(f"R:                               {r_value**2:.4f}")

print(f"\nTRABAJO DE EXTRACCIN:")
print("-" * 40)
print(f" = {trabajo_extraccion:.4e} J")
print(f" = {trabajo_extraccion_ev:.3f} eV")

print(f"\nECUACIN DEL MODELO:")
print("-" * 40)
print(f"E = ({h_experimental:.3e})f - ({abs(intercepto):.3e})")

# Tabla resumen
print(f"\nTABLA RESUMEN:")
print("-" * 40)
print(f"{'Parmetro':<25} {'Valor':<20} {'Unidades':<10}")
print("-" * 40)
print(f"{'h experimental':<25} {h_experimental:<20.4e} {'Js':<10}")
print(f"{'h terico':<25} {h_teorico:<20.4e} {'Js':<10}")
print(f"{'Error porcentual':<25} {error_porcentual:<20.2f} {'%':<10}")
print(f"{'Trabajo extraccin':<25} {trabajo_extraccion_ev:<20.3f} {'eV':<10}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Constantes fsicas
c = 3e8  # velocidad de la luz en m/s
e = 1.602e-19  # carga del electrn en C
h_teorico = 6.626e-34  # valor terico de Planck en Js

# Datos para 4mm
voltaje_4mm = np.array([-0.55, -0.74, -1.29, -1.53, -2.13])
longitud_onda_m_4mm = np.array([5.77e-7, 5.46e-7, 4.36e-7, 4.05e-7, 3.65e-7])

# Datos para 2mm
voltaje_2mm = np.array([-0.8, -0.92, -1.39, -1.53, -2.11])
longitud_onda_m_2mm = np.array([5.77e-7, 5.46e-7, 4.36e-7, 4.05e-7, 3.65e-7])

# Datos para 8mm
voltaje_8mm = np.array([-0.81, -0.89, -1.41, -1.62, -2.17])
longitud_onda_m_8mm = np.array([5.77e-7, 5.46e-7, 4.36e-7, 4.05e-7, 3.65e-7])

def calcular_planck(voltaje, longitud_onda, distancia):
    """Calcula la constante de Planck para un conjunto de datos"""
    frecuencia = c / longitud_onda
    energia = e * np.abs(voltaje)

    # Ajuste lineal
    pendiente, intercepto, r_value, p_value, std_err = stats.linregress(frecuencia, energia)
    h_experimental = pendiente

    # Clculo de errores
    error_absoluto = abs(h_experimental - h_teorico)
    error_porcentual = (error_absoluto / h_teorico) * 100

    # Trabajo de extraccin
    trabajo_extraccion = -intercepto
    trabajo_extraccion_ev = trabajo_extraccion / e

    return {
        'distancia': distancia,
        'frecuencia': frecuencia,
        'energia': energia,
        'h_experimental': h_experimental,
        'intercepto': intercepto,
        'error_porcentual': error_porcentual,
        'r_cuadrado': r_value**2,
        'trabajo_extraccion_ev': trabajo_extraccion_ev,
        'pendiente': pendiente,
        'std_err': std_err
    }

# Calcular para las tres distancias
resultados_4mm = calcular_planck(voltaje_4mm, longitud_onda_m_4mm, "4mm")
resultados_2mm = calcular_planck(voltaje_2mm, longitud_onda_m_2mm, "2mm")
resultados_8mm = calcular_planck(voltaje_8mm, longitud_onda_m_8mm, "8mm")

# Lista de todos los resultados
todos_resultados = [resultados_4mm, resultados_2mm, resultados_8mm]

# Grfico comparativo
plt.figure(figsize=(12, 8))

# Colores para cada distancia
colores = ['blue', 'red', 'green']
distancias = ['4mm', '2mm', '8mm']

for i, resultados in enumerate(todos_resultados):
    frecuencia = resultados['frecuencia']
    energia = resultados['energia']
    pendiente = resultados['pendiente']
    intercepto = resultados['intercepto']

    # Puntos experimentales
    plt.plot(frecuencia, energia, 'o', color=colores[i], markersize=8,
             label=f'{distancias[i]} - Datos')

    # Lnea de ajuste
    freq_ajuste = np.linspace(min(frecuencia), max(frecuencia), 100)
    energia_ajuste = pendiente * freq_ajuste + intercepto
    plt.plot(freq_ajuste, energia_ajuste, '-', color=colores[i], linewidth=2,
             label=f'{distancias[i]} - Ajuste (h={resultados["h_experimental"]:.3e})')

plt.xlabel('Frecuencia (Hz)', fontsize=12)
plt.ylabel('Energa (J)', fontsize=12)
plt.title('Determinacin de la Constante de Planck - Comparacin de Distancias', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()

# Resultados detallados
print("=" * 80)
print("ANLISIS COMPARATIVO DEL EFECTO FOTOELCTRICO - TRES DISTANCIAS")
print("=" * 80)

# Mostrar resultados para cada distancia
for resultados in todos_resultados:
    print(f"\n{resultados['distancia'].upper()} - RESULTADOS:")
    print("-" * 50)
    print(f"Constante de Planck experimental: {resultados['h_experimental']:.4e} Js")
    print(f"Error porcentual:                 {resultados['error_porcentual']:.2f} %")
    print(f"Trabajo de extraccin:            {resultados['trabajo_extraccion_ev']:.3f} eV")
    print(f"Coeficiente R:                   {resultados['r_cuadrado']:.4f}")
    print(f"Error estndar:                   {resultados['std_err']:.4e}")

# Clculo del valor promedio y error experimental
h_valores = [r['h_experimental'] for r in todos_resultados]
errores_porcentuales = [r['error_porcentual'] for r in todos_resultados]

h_promedio = np.mean(h_valores)
error_promedio = np.mean(errores_porcentuales)
desviacion_estandar = np.std(h_valores)
error_relativo_promedio = (desviacion_estandar / h_promedio) * 100

print(f"\n" + "=" * 80)
print("RESULTADO FINAL Y ANLISIS DE ERRORES EXPERIMENTALES")
print("=" * 80)

print(f"\nVALOR TERICO DE REFERENCIA:")
print(f"h = {h_teorico:.6e} Js")

print(f"\nVALORES EXPERIMENTALES OBTENIDOS:")
for i, resultados in enumerate(todos_resultados):
    print(f"{distancias[i]}: h = {resultados['h_experimental']:.6e} Js (Error: {resultados['error_porcentual']:.2f}%)")

print(f"\nANLISIS ESTADSTICO:")
print(f"Valor promedio de h:          {h_promedio:.6e} Js")
print(f"Desviacin estndar:          {desviacion_estandar:.6e} Js")
print(f"Error porcentual promedio:    {error_promedio:.2f} %")
print(f"Error relativo entre medidas: {error_relativo_promedio:.2f} %")

print(f"\nPRECISIN EXPERIMENTAL:")
print(f"El experimento tiene una precisin promedio de {error_promedio:.2f}%")
print(f"La consistencia entre medidas es del {100-error_relativo_promedio:.2f}%")

# Grfico de barras comparativo
plt.figure(figsize=(10, 6))
distancias_nombres = ['4mm', '2mm', '8mm']
h_experimentales = [r['h_experimental'] for r in todos_resultados]
errores = [r['std_err'] for r in todos_resultados]

x_pos = np.arange(len(distancias_nombres))
bars = plt.bar(x_pos, h_experimentales, yerr=errores, capsize=5,
               alpha=0.7, color=['blue', 'red', 'green'])

plt.axhline(y=h_teorico, color='black', linestyle='--', linewidth=2,
            label=f'h terico = {h_teorico:.3e} Js')

plt.xlabel('Distancia', fontsize=12)
plt.ylabel('Constante de Planck (Js)', fontsize=12)
plt.title('Comparacin de la Constante de Planck por Distancia', fontsize=14)
plt.xticks(x_pos, distancias_nombres)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')

# Aadir valores en las barras
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.05e-34,
             f'{h_experimentales[i]:.3e}\n({errores_porcentuales[i]:.1f}%)',
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# Tabla resumen final
print(f"\n" + "=" * 80)
print("TABLA RESUMEN FINAL")
print("=" * 80)
print(f"{'Distancia':<10} {'h_exp (Js)':<15} {'Error %':<10} {' (eV)':<10} {'R':<8}")
print("-" * 80)
for resultados in todos_resultados:
    print(f"{resultados['distancia']:<10} {resultados['h_experimental']:.3e}  {resultados['error_porcentual']:<9.2f} {resultados['trabajo_extraccion_ev']:<9.3f} {resultados['r_cuadrado']:.4f}")

print(f"\nMEJOR ESTIMACIN EXPERIMENTAL:")
print(f"h = ({h_promedio:.3e}  {desviacion_estandar:.3e}) Js")
print(f"Precisin promedio: {error_promedio:.2f}% respecto al valor terico")
```
